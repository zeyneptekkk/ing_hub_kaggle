# Kurulum & Dosyaları Oku

tipleri/dizileri doğru alalım; ileri adımlarda sürtünme olmasın.


```python
# 1) KÜTÜPHANELER
import os
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

# 2) YARDIMCI (opsiyonel): Aynı klasördeki dosya adını seçer, yoksa /mnt/data'da dener
def choose_file(*names):
    for n in names:
        if os.path.exists(n):
            return n
    for n in names:
        p = os.path.join("/mnt/data", n)
        if os.path.exists(p):
            return p
    raise FileNotFoundError(f"Şu isimlerle dosya bulunamadı: {names}")

# 3) SADE DOSYA ADLARI (aynı klasör)
HIST_FILE = choose_file("customer_history.csv")
CUST_FILE = choose_file("customers.csv")
REF_FILE  = choose_file("reference_data.csv", "referance_data.csv")          # ikisine de bak
TEST_FILE = choose_file("reference_data_test.csv", "referance_data_test.csv")
SUB_FILE  = choose_file("sample_submission.csv")

# 4) OKU
hist = pd.read_csv(HIST_FILE, parse_dates=["date"], low_memory=False)
cust = pd.read_csv(CUST_FILE, low_memory=False)
ref  = pd.read_csv(REF_FILE,  parse_dates=["ref_date"], low_memory=False)
test = pd.read_csv(TEST_FILE, parse_dates=["ref_date"], low_memory=False)
sub  = pd.read_csv(SUB_FILE,  low_memory=False)

# 5) TİP NORMALİZASYONU
for df in (hist, cust, ref, test, sub):
    if "cust_id" in df.columns:
        df["cust_id"] = pd.to_numeric(df["cust_id"], errors="coerce").astype("Int64")

if "churn" in ref.columns:
    ref["churn"] = pd.to_numeric(ref["churn"], errors="coerce").fillna(0).astype("Int8")

# 6) HIZLI KONTROL
print("hist:", hist.shape, "cust:", cust.shape, "ref:", ref.shape, "test:", test.shape, "sub:", sub.shape)
if "churn" in ref.columns:
    print("ref label balance:", ref["churn"].value_counts(normalize=True).to_dict())
```

# Pencere & Özellik Üretim Fonksiyonları

ref_date’ten önceki 1/3/6/12 aylık pencereler için özetler (sum/mean/std/max), recency, intensity (işlem başına tutar) ve trend (kısa/uzun fark & oran).

,

,

,
Her (cust_id, ref_date) için, ref_date’ten ÖNCEKİ müşteri davranışına bakıp özellikler (feature) üretiyoruz.

Bu özelliklerden bir eğitim tablosu yapıyoruz.

Sonra zamansal train/valid ayırıp hızlı bir Logistic Regression modeli ile Gini / Recall@10 / Lift@10 hesaplıyoruz.

Aynı işlemi test anahtarlarına uygulayıp submission.csv oluşturuyoruz.


.


.



L1 (çok kısa): en taze sinyal, ani düşüş/yükselişi yakalar.

L3 (kısa-orta): L1 kadar gürültülü değil; yakın dönem ortalamasını verir.

L6 (etiket ufkuyla uyumlu): son 6 ay seviyesini özetler.

L12 (uzun taban): müşterinin “normal”ini gösterir; L1/L3 ile kıyaslayınca bozulma çıkar.


-

— müşteri geçmişinden (history) sayısal özellik (feature) üretme sistemini kuruyor.
,


,


Ne yapıyor:

Her müşteri için son 1, 3, 6, 12 ay performansını özetleyeceğiz.

Bu yüzden “L1M”, “L3M”, “L6M”, “L12M” pencereleri tanımlanıyor.

BASE_NUM_COLS bu pencereler için özetlenecek sayısal kolonlar.

Neden:

Müşteri davranışını zaman içinde görmek (ör. son 3 ayda az mı işlem yapmış?)

“Trend” veya “Recency” gibi churn sinyalleri için temel oluşturmak.


,

Genel anlamda bu kod ne yapıyor?

Bu kod:

Her müşteri için geçmiş işlemleri (1,3,6,12 aylık) özetliyor,

Bu özetlerden istatistiksel özellikler (sum, mean, std, max) üretiyor,

Ek olarak recency, intensity, trend gibi türetilmiş sinyaller oluşturuyor,

Sonuçta her (cust_id, ref_date) çifti için tek satırlık model girişi hazırlıyor.

```python
# Sadece lazım kolonları tut (emniyet)
HIST_COLS = [
    "cust_id","date",
    "mobile_eft_all_cnt","active_product_category_nbr",
    "mobile_eft_all_amt","cc_transaction_all_amt","cc_transaction_all_cnt"
]
hist = hist[[c for c in HIST_COLS if c in hist.columns]].copy()
hist = hist.sort_values(["cust_id","date"]).reset_index(drop=True)

WINDOWS = [1,3,6,12]   # L1/L3/L6/L12
BASE_NUM_COLS = [c for c in HIST_COLS if c not in ["cust_id","date"]]

def _agg_safe(series, fn):
    """Boş/tek elemanlı serilerde güvenli özet."""
    s = pd.to_numeric(series, errors="coerce")
    s = s[np.isfinite(s)]
    if s.size == 0:
        if fn == "sum":  return 0.0
        if fn == "mean": return 0.0
        if fn == "std":  return 0.0
        if fn == "max":  return 0.0
    if fn == "sum":  return float(s.sum())
    if fn == "mean": return float(s.mean())
    if fn == "std":  return float(s.std(ddof=0))   # ddof=0 -> tek elemanda 0
    if fn == "max":  return float(s.max())
    return 0.0

def aggregate_for_one(h_cust: pd.DataFrame, ref_date: pd.Timestamp, windows=WINDOWS):
    """
    Tek bir (cust_id, ref_date) için özellikleri üret.
    Sadece date < ref_date olan geçmişe bakar.
    """
    feat = {"ref_date": ref_date}
    h = h_cust[h_cust["date"] < ref_date]
    if h.empty:
        for col in BASE_NUM_COLS:
            for w in windows:
                feat[f"{col}_L{w}M_sum"]  = 0.0
                feat[f"{col}_L{w}M_mean"] = 0.0
                feat[f"{col}_L{w}M_std"]  = 0.0
                feat[f"{col}_L{w}M_max"]  = 0.0
        feat["recency_days"] = 9999
    else:
        feat["recency_days"] = int((ref_date - h["date"].max()).days)
        for w in windows:
            start = ref_date - pd.DateOffset(months=w)
            hw = h[h["date"] >= start]
            for col in BASE_NUM_COLS:
                feat[f"{col}_L{w}M_sum"]  = _agg_safe(hw[col], "sum")
                feat[f"{col}_L{w}M_mean"] = _agg_safe(hw[col], "mean")
                feat[f"{col}_L{w}M_std"]  = _agg_safe(hw[col], "std")
                feat[f"{col}_L{w}M_max"]  = _agg_safe(hw[col], "max")

    # INTENSITY: tutar / adet
    eps = 1e-6
    for w in windows:
        amt = feat.get(f"cc_transaction_all_amt_L{w}M_sum", 0.0)
        cnt = feat.get(f"cc_transaction_all_cnt_L{w}M_sum", 0.0)
        feat[f"cc_amt_per_txn_L{w}M"] = float(amt / (cnt + eps))

        meft_amt = feat.get(f"mobile_eft_all_amt_L{w}M_sum", 0.0)
        meft_cnt = feat.get(f"mobile_eft_all_cnt_L{w}M_sum", 0.0)
        feat[f"mobile_eft_amt_per_txn_L{w}M"] = float(meft_amt / (meft_cnt + eps))

    # TREND: kısa vs uzun (mean bazlı diff & ratio)
    for short, long in [(1,3),(3,6),(6,12)]:
        for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt","active_product_category_nbr"]:
            mS = feat.get(f"{base}_L{short}M_mean", 0.0)
            mL = feat.get(f"{base}_L{long}M_mean",  0.0)
            feat[f"{base}_trend_mean_L{short}vL{long}_diff"]  = float(mS - mL)
            feat[f"{base}_trend_mean_L{short}vL{long}_ratio"] = float(mS / (mL + 1e-6))

    return feat

def build_features_for_keys(hist_df, keys_df, n_limit=None):
    """
    keys_df: ['cust_id','ref_date', ('churn' varsa ekleriz)]
    n_limit: ilk N anahtarda çalış (prototip için)
    """
    if n_limit is not None:
        keys_df = keys_df.head(n_limit).copy()

    # müşteri bazında cache: her anahtar için yeniden filtrelemeyelim
    hist_df = hist_df.sort_values(["cust_id","date"])
    cache = {}
    rows = []
    for _, row in keys_df.iterrows():
        cid = row["cust_id"]; ref_date = row["ref_date"]
        if pd.isna(cid) or pd.isna(ref_date): 
            continue
        if cid in cache:
            h_cust = cache[cid]
        else:
            h_cust = hist_df[hist_df["cust_id"] == cid]
            cache[cid] = h_cust
        feat = aggregate_for_one(h_cust, ref_date)
        feat["cust_id"] = cid
        rows.append(feat)
    return pd.DataFrame(rows)
```

# 3) Eğitim Özellikleri (küçük prototip) + Demografiler

: Önce küçük bir örnekle hattı doğrulayalım; sonra tüm veriye uygularız.

1) Zaman penceresi filtresi (temel kural)

Her (cust_id, ref_date) için:

Yalnızca date < ref_date olan geçmiş satırlar kullanılır (leakage yok).

Pencere başlangıcı: start = ref_date − {1,3,6,12} ay

O pencerenin verisi: start ≤ date < ref_date



.
referance_data.csv dosyasından ilk 1000 sutunu alıp modeli eğitiyor ve  200+ sütunluk özellik tablosu döner

```python
# Küçük prototip: ilk 1,000 etiketli anahtar
N_KEYS = 1000
train_keys = ref[["cust_id","ref_date","churn"]].copy()

X_hist_small = build_features_for_keys(hist, train_keys[["cust_id","ref_date"]], n_limit=N_KEYS)

# Join: labels + demographics
X_small = train_keys.head(N_KEYS).merge(X_hist_small, on=["cust_id","ref_date"], how="left")
X_small = X_small.merge(cust, on="cust_id", how="left")

# Eksik doldurma
num_cols = X_small.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X_small.columns if c not in num_cols and c not in ["ref_date"]]

X_small[num_cols] = X_small[num_cols].fillna(0)
for c in cat_cols:
    X_small[c] = X_small[c].fillna("Unknown")

print("X_small shape:", X_small.shape)
X_small.head(3)
```

# One-Hot Encoding + Zamansal Train/Valid Bölme

Resmi metrike benzeyen validasyon için zamana göre böl.

,

Bu kodun görevi:

✅ veriyi makine öğrenmesi modeline uygun hale getirmek
✅ zamana göre train / validation (eğitim ve doğrulama) setine bölmek
,


💬 kısacası:

bu kod bloğu, veriyi modelin anlayacağı sayısal formata dönüştürüp
geçmişe göre eğitim / test kümelerine ayırıyor.


```python
TARGET = "churn"
PROTECTED = ["cust_id", "ref_date", TARGET]

# 1) Kategorikleri bul (object tipli ve protected olmayanlar)
cat_cols = [c for c in X_small.columns if (X_small[c].dtype == "object") and (c not in PROTECTED)]

# 2) One-Hot Encoding (sayılara çeviriyoruz)
X_enc = pd.get_dummies(X_small, columns=cat_cols, drop_first=True)

# 3) Zamansal eşik: ref_date'in %80'lik alt dilim tarihi
cut = X_enc["ref_date"].quantile(0.80)

# 4) Maskeler: erken tarihler train, geç tarihler valid
train_mask = X_enc["ref_date"] < cut
valid_mask = ~train_mask

# 5) Hedef ve özellik matrisleri
y_train = X_enc.loc[train_mask, TARGET].astype(int).values
y_valid = X_enc.loc[valid_mask, TARGET].astype(int).values
# year-month sütunu yap
ym = X_enc["ref_date"].dt.to_period("M")
#

feat_cols = [c for c in X_enc.columns if c not in PROTECTED]
X_train = X_enc.loc[train_mask, feat_cols].fillna(0)
X_valid = X_enc.loc[valid_mask, feat_cols].fillna(0)

print("cut date  :", cut)
print("train size:", X_train.shape, "| valid size:", X_valid.shape)
print("valid min/max ref_date:", X_enc.loc[valid_mask, "ref_date"].min(), X_enc.loc[valid_mask, "ref_date"].max())
```

# Baseline Model (Logistic, class_weight='balanced') + Yarışma Metriği

hızlı, sağlam bir başlangıç ve resmi skora benzer ölçümi

,
lojistik regresyon modeli kullandık burda

,

💬 Kısacası:

Bu kod, modelini eğitip performansını üç farklı açıdan ölçüyor
ve yarışmadaki puanlama sistemine göre bir “final skor” üretiyor.


```python
# LOGISTIC
clf = LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear")
clf.fit(X_train, y_train)
valid_proba = clf.predict_proba(X_valid)[:,1]

# METRİKLER
def gini_from_auc(y_true, y_score):
    auc = roc_auc_score(y_true, y_score)
    return 2*auc - 1

def recall_at_k(y_true, y_score, k=0.10):
    n = len(y_true); topk = int(np.ceil(n*k))
    idx = np.argsort(-y_score)[:topk]
    return float(y_true[idx].sum()) / float(y_true.sum() + 1e-9)

def lift_at_k(y_true, y_score, k=0.10):
    n = len(y_true); topk = int(np.ceil(n*k))
    idx = np.argsort(-y_score)[:topk]
    precision_k = float(y_true[idx].mean())
    prevalence  = float(y_true.mean())
    return float(precision_k / max(prevalence, 1e-9))

gini  = gini_from_auc(y_valid, valid_proba)
rec10 = recall_at_k(y_valid, valid_proba, 0.10)
lift10= lift_at_k(y_valid, valid_proba, 0.10)
score = 0.4*gini + 0.3*rec10 + 0.3*lift10

print({"Gini":gini, "Recall@10%":rec10, "Lift@10%":lift10, "CompetitionScore":score})
```

bu skorlar (Gini≈0.14, Recall@10≈0.045, Lift@10≈0.446) baseline’ın zayıf olduğunu söylüyor.bu başlangıç için normal 

# Tüm Eğitim İçin Özellik Üret

Pipeline doğrulandıysa, şimdi tüm etiketli veriye uygula.

,

Bu kısım, küçük örnekte doğruladığımız pipeline’ı bütün eğitim verisine uygular.

```python
# (İSTERSEN) TÜM EĞİTİM
# UYARI: Daha büyük veri. Gerekirse n_limit=None ile tam çalışır.
FULL = False  # True yaparsan tüm eğitim anahtarlarını işler

if FULL:
    X_hist_full = build_features_for_keys(hist, ref[["cust_id","ref_date"]], n_limit=None)
    X_full = ref.merge(X_hist_full, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")

    num_cols = X_full.select_dtypes(include=[np.number]).columns.tolist()
    cat_cols = [c for c in X_full.columns if c not in num_cols and c not in ["ref_date"]]
    X_full[num_cols] = X_full[num_cols].fillna(0)
    for c in cat_cols:
        X_full[c] = X_full[c].fillna("Unknown")

    # Encode
    X_full_enc = pd.get_dummies(X_full, columns=cat_cols, drop_first=True)

    # Zamansal bölme
    cut = X_full_enc["ref_date"].quantile(0.80)
    train_mask = X_full_enc["ref_date"] < cut
    valid_mask = ~train_mask

    y_train = X_full_enc.loc[train_mask, "churn"].astype(int).values
    y_valid = X_full_enc.loc[valid_mask, "churn"].astype(int).values

    feat_cols = [c for c in X_full_enc.columns if c not in ["cust_id","ref_date","churn"]]
    X_train = X_full_enc.loc[train_mask, feat_cols].fillna(0)
    X_valid = X_full_enc.loc[valid_mask, feat_cols].fillna(0)

    clf = LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear")
    clf.fit(X_train, y_train)
    valid_proba = clf.predict_proba(X_valid)[:,1]

    gini  = gini_from_auc(y_valid, valid_proba)
    rec10 = recall_at_k(y_valid, valid_proba, 0.10)
    lift10= lift_at_k(y_valid, valid_proba, 0.10)
    score = 0.4*gini + 0.3*rec10 + 0.3*lift10

    print({"Gini":gini, "Recall@10%":rec10, "Lift@10%":lift10, "CompetitionScore":score})

    # Eğitim tamamlandıktan sonra, test için de aynı encoder kolonları lazım:
    TRAIN_FEATS = X_train.columns.tolist()
```

```python
# === (A) TÜM EĞİTİM ÖZELLİKLERİNİ PARTİ PARTİ ÜRET ===
# varsayım: build_features_for_keys, aggregate_for_one fonksiyonların hazır

import pandas as pd

BATCH = 20000  # her partide işlenecek anahtar sayısı (donanımına göre artır/azalt)
keys = ref[["cust_id","ref_date","churn"]].sort_values("ref_date").reset_index(drop=True)

parts = []
for s in range(0, len(keys), BATCH):
    e = min(s + BATCH, len(keys))
    keys_part = keys.iloc[s:e][["cust_id","ref_date"]]
    feats_part = build_features_for_keys(hist, keys_part, n_limit=None)
    # etiket ve demografi ekle
    part = keys.iloc[s:e].merge(feats_part, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
    parts.append(part)
    print(f"done: {e}/{len(keys)}")

X_full = pd.concat(parts, ignore_index=True)

# eksik doldur
num_cols = X_full.select_dtypes(include="number").columns
cat_cols = [c for c in X_full.columns if c not in num_cols and c != "ref_date"]
X_full[num_cols] = X_full[num_cols].fillna(0)
for c in cat_cols: X_full[c] = X_full[c].fillna("Unknown")
```

```python
# === (B) ENCODE + ZAMANSAL SPLIT + MODEL + METRİK ===
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
import numpy as np

TARGET = "churn"
PROTECTED = ["cust_id","ref_date",TARGET]

# one-hot
X_full_enc = pd.get_dummies(X_full, columns=[c for c in X_full.columns if (X_full[c].dtype=="object" and c not in PROTECTED)], drop_first=True)

# zamansal bölme — dilersen son 6 ayı valid yapabilirsin
cut = X_full_enc["ref_date"].quantile(0.80)
train_mask = X_full_enc["ref_date"] < cut
valid_mask = ~train_mask

y_train = X_full_enc.loc[train_mask, TARGET].astype(int).values
y_valid = X_full_enc.loc[valid_mask, TARGET].astype(int).values

feat_cols = [c for c in X_full_enc.columns if c not in PROTECTED]
X_train = X_full_enc.loc[train_mask, feat_cols].fillna(0)
X_valid = X_full_enc.loc[valid_mask,  feat_cols].fillna(0)

# Logistic baseline
clf = LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear")
clf.fit(X_train, y_train)
valid_proba = clf.predict_proba(X_valid)[:,1]

# metrikler
def gini_from_auc(y_true, y_score): return 2*roc_auc_score(y_true, y_score) - 1
def recall_at_k(y_true, y_score, k=0.10):
    n=len(y_true); topk=int(np.ceil(n*k)); idx=np.argsort(-y_score)[:topk]
    return float(y_true[idx].sum())/float(y_true.sum()+1e-9)
def lift_at_k(y_true, y_score, k=0.10):
    n=len(y_true); topk=int(np.ceil(n*k)); idx=np.argsort(-y_score)[:topk]
    prec=float(y_true[idx].mean()); prev=float(y_true.mean())
    return prec/max(prev,1e-9)

gini  = gini_from_auc(y_valid, valid_proba)
rec10 = recall_at_k(y_valid, valid_proba, 0.10)
lift10= lift_at_k(y_valid, valid_proba, 0.10)
score = 0.4*gini + 0.3*rec10 + 0.3*lift10
print({"Gini":gini, "Recall@10%":rec10, "Lift@10%":lift10, "CompetitionScore":score})

# eğitimdeki sütun düzenini kaydet (test'i hizalamak için)
TRAIN_FEATS = X_train.columns.tolist()
```

# Test Özellikleri + Submission

```python
# === Test pipeline + submission (yol otomatik bulma, gerekirse yeniden eğitim) ===
import os, sys, glob
import pandas as pd
import numpy as np

# ---------- 0) Yardımcı: dosya bulucu ----------
def find_file(fname: str, search_roots=(".",)):
    """fname'i verilen köklerde (recursive) arar. İlk bulunan yolu döner, yoksa None."""
    fname = fname.lower()
    for root in search_roots:
        for path in glob.iglob(os.path.join(root, "**", "*"), recursive=True):
            if os.path.isfile(path) and os.path.basename(path).lower() == fname:
                return os.path.abspath(path)
    return None

def resolve_path(default_path: str, fallback_name: str):
    """Varsayılan path varsa onu, yoksa çalışma klasöründe dosya adını arar."""
    if os.path.exists(default_path):
        return default_path
    p = find_file(fallback_name, search_roots=(".", os.getcwd()))
    if p: 
        print(f"[info] Bulundu: {fallback_name} -> {p}")
        return p
    raise FileNotFoundError(f"Dosya bulunamadı: {fallback_name}. Lütfen yolu ayarla.")

# ---------- 1) Dosyaları bul/oku ----------
PATH_HISTORY = resolve_path("/mnt/data/customer_history.csv", "customer_history.csv")
PATH_CUSTOMERS = resolve_path("/mnt/data/customers.csv", "customers.csv")
PATH_REF = resolve_path("/mnt/data/referance_data.csv", "referance_data.csv")  # yarışma dosya adı 'referance'
PATH_TEST = resolve_path("/mnt/data/referance_data_test.csv", "referance_data_test.csv")
PATH_SUB = resolve_path("/mnt/data/sample_submission.csv", "sample_submission.csv")

hist = pd.read_csv(PATH_HISTORY, parse_dates=["date"], low_memory=False)
cust = pd.read_csv(PATH_CUSTOMERS, low_memory=False)
ref  = pd.read_csv(PATH_REF, parse_dates=["ref_date"], low_memory=False)
test = pd.read_csv(PATH_TEST, parse_dates=["ref_date"], low_memory=False)
sub  = pd.read_csv(PATH_SUB, low_memory=False)

# tipler
for df in (hist, cust, ref, test, sub):
    if "cust_id" in df.columns:
        df["cust_id"] = pd.to_numeric(df["cust_id"], errors="coerce").astype("Int64")
if "churn" in ref.columns:
    ref["churn"] = pd.to_numeric(ref["churn"], errors="coerce").fillna(0).astype("Int8")

# ---------- 2) Özellik fonksiyonları (eğitimdekilerle birebir) ----------
HIST_COLS = [
    "cust_id","date",
    "mobile_eft_all_cnt","active_product_category_nbr",
    "mobile_eft_all_amt","cc_transaction_all_amt","cc_transaction_all_cnt"
]
hist = hist[[c for c in HIST_COLS if c in hist.columns]].sort_values(["cust_id","date"]).reset_index(drop=True)
WINDOWS = [1,3,6,12]
BASE_NUM_COLS = [c for c in HIST_COLS if c not in ["cust_id","date"]]

def _agg_safe(s, fn):
    s = pd.to_numeric(s, errors="coerce"); s = s[np.isfinite(s)]
    if s.size == 0: return 0.0
    if fn=="sum":  return float(s.sum())
    if fn=="mean": return float(s.mean())
    if fn=="std":  return float(s.std(ddof=0))
    if fn=="max":  return float(s.max())
    return 0.0

def aggregate_for_one(h_cust: pd.DataFrame, ref_date: pd.Timestamp):
    feat = {"ref_date": ref_date}
    h = h_cust[h_cust["date"] < ref_date]
    if h.empty:
        for col in BASE_NUM_COLS:
            for w in WINDOWS:
                feat[f"{col}_L{w}M_sum"]=feat[f"{col}_L{w}M_mean"]=feat[f"{col}_L{w}M_std"]=feat[f"{col}_L{w}M_max"]=0.0
        feat["recency_days"] = 9999
    else:
        feat["recency_days"] = int((ref_date - h["date"].max()).days)
        for w in WINDOWS:
            start = ref_date - pd.DateOffset(months=w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            for col in BASE_NUM_COLS:
                feat[f"{col}_L{w}M_sum"]  = _agg_safe(hw[col], "sum")
                feat[f"{col}_L{w}M_mean"] = _agg_safe(hw[col], "mean")
                feat[f"{col}_L{w}M_std"]  = _agg_safe(hw[col], "std")
                feat[f"{col}_L{w}M_max"]  = _agg_safe(hw[col], "max")
    # intensity
    eps=1e-6
    for w in WINDOWS:
        feat[f"cc_amt_per_txn_L{w}M"] = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0) / (feat.get(f"cc_transaction_all_cnt_L{w}M_sum",0.0)+eps)
        feat[f"mobile_eft_amt_per_txn_L{w}M"] = feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0) / (feat.get(f"mobile_eft_all_cnt_L{w}M_sum",0.0)+eps)
    # trend
    for short,long in [(1,3),(3,6),(6,12)]:
        for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt","active_product_category_nbr"]:
            ms = feat.get(f"{base}_L{short}M_mean",0.0); ml = feat.get(f"{base}_L{long}M_mean",0.0)
            feat[f"{base}_trend_mean_L{short}vL{long}_diff"]  = ms-ml
            feat[f"{base}_trend_mean_L{short}vL{long}_ratio"] = ms/(ml+1e-6)
    return feat

def build_features_for_keys(hist_df, keys_df, n_limit=None, progress_every=5000):
    if n_limit is not None:
        keys_df = keys_df.head(n_limit).copy()
    h = hist_df  # zaten filtreli/sıralı
    cache = {}
    rows = []
    for i, r in enumerate(keys_df.itertuples(index=False), 1):
        cid, rd = r.cust_id, r.ref_date
        if pd.isna(cid) or pd.isna(rd): 
            continue
        if cid in cache: h_cust = cache[cid]
        else:
            h_cust = h[h["cust_id"] == cid]; cache[cid] = h_cust
        f = aggregate_for_one(h_cust, rd); f["cust_id"]=cid; rows.append(f)
        if progress_every and (i % progress_every == 0):
            print(f"[features] {i}/{len(keys_df)}")
    return pd.DataFrame(rows)

# ---------- 3) (Varsa) eğitimden gelen model/sütunlar; yoksa yeniden eğit ----------
retrain_needed = False
try:
    TRAIN_FEATS  # eğitimde kaydettiysen burada vardır
    clf          # eğitimde fit ettiğin model
    _ = TRAIN_FEATS[0]
    print("[info] Eğitimden TRAIN_FEATS ve clf bulundu (yeniden eğitim yok).")
except Exception:
    print("[info] TRAIN_FEATS/clf yok; basit Logistic ile yeniden eğitim yapılacak.")
    retrain_needed = True

if retrain_needed:
    # Tüm eğitim anahtarlarını parça parça özelliklendirelim (sen daha önce yaptın)
    BATCH = 20000
    keys = ref[["cust_id","ref_date","churn"]].sort_values("ref_date").reset_index(drop=True)
    parts = []
    for s in range(0, len(keys), BATCH):
        e = min(s + BATCH, len(keys))
        keys_part = keys.iloc[s:e][["cust_id","ref_date"]]
        feats_part = build_features_for_keys(hist, keys_part, n_limit=None)
        part = keys.iloc[s:e].merge(feats_part, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
        parts.append(part)
        print(f"[train] done: {e}/{len(keys)}")
    X_full = pd.concat(parts, ignore_index=True)

    # eksik doldur
    num_cols = X_full.select_dtypes(include="number").columns
    cat_cols = [c for c in X_full.columns if c not in num_cols and c != "ref_date"]
    X_full[num_cols] = X_full[num_cols].fillna(0)
    for c in cat_cols: X_full[c] = X_full[c].fillna("Unknown")

    # encode
    PROTECTED = ["cust_id","ref_date","churn"]
    X_full_enc = pd.get_dummies(X_full, columns=[c for c in X_full.columns if (X_full[c].dtype=="object" and c not in PROTECTED)], drop_first=True)

    # zamansal split (80/20)
    cut = X_full_enc["ref_date"].quantile(0.80)
    train_mask = X_full_enc["ref_date"] < cut
    valid_mask = ~train_mask

    from sklearn.linear_model import LogisticRegression
    from sklearn.metrics import roc_auc_score

    y_train = X_full_enc.loc[train_mask, "churn"].astype(int).values
    y_valid = X_full_enc.loc[valid_mask, "churn"].astype(int).values
    feat_cols = [c for c in X_full_enc.columns if c not in PROTECTED]
    X_train = X_full_enc.loc[train_mask, feat_cols].fillna(0)
    X_valid = X_full_enc.loc[valid_mask,  feat_cols].fillna(0)

    clf = LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear")
    clf.fit(X_train, y_train)

    # metrik yazdır (isteğe bağlı)
    valid_proba = clf.predict_proba(X_valid)[:,1]
    def gini_from_auc(y_true, y_score): return 2*roc_auc_score(y_true, y_score) - 1
    def recall_at_k(y_true, y_score, k=0.10):
        n=len(y_true); topk=int(np.ceil(n*k)); idx=np.argsort(-y_score)[:topk]
        return float(y_true[idx].sum())/float(y_true.sum()+1e-9)
    def lift_at_k(y_true, y_score, k=0.10):
        n=len(y_true); topk=int(np.ceil(n*k)); idx=np.argsort(-y_score)[:topk]
        prec=float(y_true[idx].mean()); prev=float(y_true.mean())
        return prec/max(prev,1e-9)
    gini  = gini_from_auc(y_valid, valid_proba)
    rec10 = recall_at_k(y_valid, valid_proba, 0.10)
    lift10= lift_at_k(y_valid, valid_proba, 0.10)
    score = 0.4*gini + 0.3*rec10 + 0.3*lift10
    print({"Gini":gini, "Recall@10%":rec10, "Lift@10%":lift10, "CompetitionScore":score})

    TRAIN_FEATS = X_train.columns.tolist()
    print("[info] Yeniden eğitim tamam. TRAIN_FEATS ve clf hazır.")

# ---------- 4) Test özellikleri ----------
test_keys = test[["cust_id","ref_date"]].copy()
X_hist_test = build_features_for_keys(hist, test_keys, n_limit=None, progress_every=5000)

X_test = test_keys.merge(X_hist_test, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
num_cols = X_test.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in X_test.columns if c not in num_cols and c not in ["ref_date"]]
X_test[num_cols] = X_test[num_cols].fillna(0)
for c in cat_cols: X_test[c] = X_test[c].fillna("Unknown")

X_test_enc = pd.get_dummies(X_test, columns=cat_cols, drop_first=True)
X_test_enc = X_test_enc.reindex(columns=TRAIN_FEATS, fill_value=0)  # eğitimle hizala

# ---------- 5) Tahmin + Submission ----------
test_proba = clf.predict_proba(X_test_enc)[:, 1]
out = sub[["cust_id"]].merge(
    pd.DataFrame({"cust_id": X_test["cust_id"], "churn": test_proba}),
    on="cust_id", how="left"
)
overall_prev = float(ref["churn"].mean())
out["churn"] = out["churn"].fillna(overall_prev)
assert set(out.columns) == {"cust_id","churn"}
assert len(out) == len(sub)
assert out["churn"].between(0,1).all()

out_path = os.path.abspath("submission.csv")
out.to_csv(out_path, index=False, float_format="%.6f")
print(f"[OK] submission yazıldı: {out_path}  şekil: {out.shape}")
print(out.head(3).to_string(index=False))
```

```python
# ============================================================
# 🧠 ING Datathon — MINI (back-to-best alignment)
#   - Recent-focused sampling: SAMPLE_N=12000, RECENT_FRAC=0.55
#   - Güçlü feature'lar: last_active / zero_months / HHI / CV / EWM (span6)
#   - Gürültü azaltma: low-variance drop + gain tabanlı %15 pruning (opsiyonel)
#   - class_weight="balanced" (scale_pos_weight kapalı)
#   - Submission YOK (mini validasyon)
# ============================================================

# === 0) KÜTÜPHANELER ===
import os, glob, gc
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier, early_stopping, log_evaluation
from sklearn.metrics import roc_auc_score

pd.set_option("display.width", 160)
pd.set_option("display.max_columns", 220)

SEED = 42

# === 1) MINI PARAMS (en iyi koşuya hizalı) ===
SAMPLE_N     = 12000
FOCUS_RECENT = True
RECENT_FRAC  = 0.55

DO_LOWVAR_DROP = True
DO_PRUNE       = True
PRUNE_FRAC     = 0.15   # alt %15'i at

# === 2) DOSYA YÜKLEME ===
def find_file(fname: str, search_roots=(".",)):
    fname = fname.lower()
    for root in search_roots:
        for path in glob.iglob(os.path.join(root, "**", "*"), recursive=True):
            if os.path.isfile(path) and os.path.basename(path).lower() == fname:
                return os.path.abspath(path)
    return None

def resolve_path(default_path: str, fallback_name: str):
    if os.path.exists(default_path): return default_path
    p = find_file(fallback_name, search_roots=(".", os.getcwd()))
    if p:
        print(f"[info] Bulundu: {fallback_name} -> {p}")
        return p
    raise FileNotFoundError(f"Dosya bulunamadı: {fallback_name}")

PATH_HISTORY = resolve_path("/mnt/data/customer_history.csv", "customer_history.csv")
PATH_CUSTOMERS = resolve_path("/mnt/data/customers.csv", "customers.csv")
PATH_REF      = resolve_path("/mnt/data/referance_data.csv", "referance_data.csv")

hist = pd.read_csv(PATH_HISTORY, parse_dates=["date"], low_memory=False)
cust = pd.read_csv(PATH_CUSTOMERS, low_memory=False)
ref  = pd.read_csv(PATH_REF, parse_dates=["ref_date"], low_memory=False)

for df in (hist, cust, ref):
    if "cust_id" in df.columns:
        df["cust_id"] = pd.to_numeric(df["cust_id"], errors="coerce").astype("Int64")
if "churn" in ref.columns:
    ref["churn"] = pd.to_numeric(ref["churn"], errors="coerce").fillna(0).astype("Int8")
cust.columns = cust.columns.str.strip()
hist.columns = hist.columns.str.strip()

print("✅ Dosyalar yüklendi: ",
      "\n  hist:", hist.shape,
      "| cust:", cust.shape,
      "| ref:", ref.shape)

# === 3) FEATURE’LAR (sızıntısız) ===
HIST_COLS = [
    "cust_id","date",
    "mobile_eft_all_cnt","active_product_category_nbr",
    "mobile_eft_all_amt","cc_transaction_all_amt","cc_transaction_all_cnt"
]
hist = hist[[c for c in HIST_COLS if c in hist.columns]].sort_values(["cust_id","date"]).reset_index(drop=True)

WINDOWS = [1,3,6,12]
BASE_NUM_COLS = [c for c in HIST_COLS if c not in ["cust_id","date"]]

def _agg_safe(s, fn):
    s = pd.to_numeric(s, errors="coerce"); s = s[np.isfinite(s)]
    if s.size == 0: return 0.0
    if fn=="sum":  return float(s.sum())
    if fn=="mean": return float(s.mean())
    if fn=="std":  return float(s.std(ddof=0))
    if fn=="max":  return float(s.max())
    return 0.0

def _months_active(frame: pd.DataFrame) -> int:
    if frame.empty: return 0
    cond = (pd.to_numeric(frame.get("cc_transaction_all_cnt", 0), errors="coerce").fillna(0) > 0) | \
           (pd.to_numeric(frame.get("mobile_eft_all_cnt", 0), errors="coerce").fillna(0) > 0)
    return int(cond.sum())

def _ewm_last(series: pd.Series, span: int):
    series = pd.to_numeric(series, errors="coerce").fillna(0.0)
    if series.empty: return 0.0
    return float(series.ewm(span=span, adjust=False).mean().iloc[-1])

def aggregate_for_one(h_cust: pd.DataFrame, ref_date: pd.Timestamp):
    feat = {"ref_date": ref_date}
    h = h_cust[h_cust["date"] < ref_date]

    if h.empty:
        for col in BASE_NUM_COLS:
            for w in WINDOWS:
                feat[f"{col}_L{w}M_sum"]=feat[f"{col}_L{w}M_mean"]=feat[f"{col}_L{w}M_std"]=feat[f"{col}_L{w}M_max"]=0.0
        feat["recency_days"] = 9999
        for w in WINDOWS:
            feat[f"months_active_L{w}M"]=0.0
            feat[f"activity_rate_L{w}M"]=0.0
            feat[f"inactivity_ratio_L{w}M"]=1.0
            feat[f"cc_share_amt_L{w}M"]=0.0
            feat[f"eft_share_amt_L{w}M"]=0.0
        # EWM (yalnızca span6)
        feat["cc_amt_ewm_span6"]=0.0; feat["eft_amt_ewm_span6"]=0.0
        feat["cc_cnt_ewm_span6"]=0.0; feat["eft_cnt_ewm_span6"]=0.0
        # last active & zero months
        feat["last_active_days"]=9999; feat["last_cc_days"]=9999; feat["last_eft_days"]=9999
        feat["zero_cc_months_L3"]=3.0; feat["zero_eft_months_L3"]=3.0
        feat["zero_cc_months_L6"]=6.0; feat["zero_eft_months_L6"]=6.0
    else:
        feat["recency_days"] = int((ref_date - h["date"].max()).days)
        for w in WINDOWS:
            start = ref_date - pd.DateOffset(months=w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            for col in BASE_NUM_COLS:
                feat[f"{col}_L{w}M_sum"]  = _agg_safe(hw[col], "sum")
                feat[f"{col}_L{w}M_mean"] = _agg_safe(hw[col], "mean")
                feat[f"{col}_L{w}M_std"]  = _agg_safe(hw[col], "std")
                feat[f"{col}_L{w}M_max"]  = _agg_safe(hw[col], "max")
            ma = float(_months_active(hw))
            feat[f"months_active_L{w}M"]=ma
            feat[f"activity_rate_L{w}M"]=ma/float(w)
            feat[f"inactivity_ratio_L{w}M"]=float(max(w-ma,0))/float(w)
            eps=1e-6
            cc_sum = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0)
            eft_sum= feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0)
            tot = cc_sum + eft_sum
            feat[f"cc_share_amt_L{w}M"]=cc_sum/(tot+eps)
            feat[f"eft_share_amt_L{w}M"]=eft_sum/(tot+eps)

        # EWM (sade: span6)
        h_sorted = h.sort_values("date")
        feat["cc_amt_ewm_span6"] = _ewm_last(h_sorted["cc_transaction_all_amt"], 6)
        feat["eft_amt_ewm_span6"] = _ewm_last(h_sorted["mobile_eft_all_amt"], 6)
        feat["cc_cnt_ewm_span6"]  = _ewm_last(h_sorted["cc_transaction_all_cnt"], 6)
        feat["eft_cnt_ewm_span6"] = _ewm_last(h_sorted["mobile_eft_all_cnt"], 6)

        # last active (gerçek işlem)
        def _last_days(mask):
            if mask.any():
                last_dt = h.loc[mask, "date"].max()
                return int((ref_date - last_dt).days)
            return 9999
        cc_cnt = pd.to_numeric(h.get("cc_transaction_all_cnt", 0), errors="coerce").fillna(0)
        eft_cnt= pd.to_numeric(h.get("mobile_eft_all_cnt", 0), errors="coerce").fillna(0)
        any_act = (cc_cnt>0) | (eft_cnt>0)
        feat["last_active_days"]=_last_days(any_act)
        feat["last_cc_days"]=_last_days(cc_cnt>0)
        feat["last_eft_days"]=_last_days(eft_cnt>0)

        # zero months (3/6)
        def _zero_months(last_w):
            start = ref_date - pd.DateOffset(months=last_w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            ccz = (pd.to_numeric(hw.get("cc_transaction_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            eftz= (pd.to_numeric(hw.get("mobile_eft_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            return float(ccz), float(eftz)
        z3_cc,z3_eft=_zero_months(3); z6_cc,z6_eft=_zero_months(6)
        feat["zero_cc_months_L3"]=z3_cc; feat["zero_eft_months_L3"]=z3_eft
        feat["zero_cc_months_L6"]=z6_cc; feat["zero_eft_months_L6"]=z6_eft

    # işlem başına tutar
    eps=1e-6
    for w in WINDOWS:
        feat[f"cc_amt_per_txn_L{w}M"] = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0)/(feat.get(f"cc_transaction_all_cnt_L{w}M_sum",0.0)+eps)
        feat[f"mobile_eft_amt_per_txn_L{w}M"] = feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0)/(feat.get(f"mobile_eft_all_cnt_L{w}M_sum",0.0)+eps)

    # trendler (mean bazlı)
    for short,long in [(1,3),(3,6),(6,12)]:
        for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt","active_product_category_nbr"]:
            ms = feat.get(f"{base}_L{short}M_mean",0.0)
            ml = feat.get(f"{base}_L{long}M_mean",0.0)
            feat[f"{base}_trend_mean_L{short}vL{long}_diff"]=ms-ml
            feat[f"{base}_trend_mean_L{short}vL{long}_ratio"]=ms/(ml+1e-6)

    # kanal yoğunlaşması (HHI) & oynaklık (CV)
    for w in WINDOWS:
        cc_share  = feat.get(f"cc_share_amt_L{w}M",0.0)
        eft_share = feat.get(f"eft_share_amt_L{w}M",0.0)
        feat[f"channel_hhi_amt_L{w}M"]=cc_share**2 + eft_share**2

    for base in ["cc_transaction_all_amt","mobile_eft_all_amt","cc_transaction_all_cnt","mobile_eft_all_cnt"]:
        for w in [3,6,12]:
            stdv = feat.get(f"{base}_L{w}M_std",0.0)
            mean = feat.get(f"{base}_L{w}M_mean",0.0)
            feat[f"{base}_cv_L{w}M"]= stdv/(mean+eps)

    # recency dönüşüm
    feat["recency_log"]   = float(np.log1p(feat.get("recency_days",9999)))
    feat["recency_score"] = float(np.exp(-feat.get("recency_days",9999)/30.0))

    return feat

def build_features_for_keys(hist_df, keys_df, progress_every=3000):
    rows, cache = [], {}
    for i, r in enumerate(keys_df.itertuples(index=False), 1):
        cid, rd = r.cust_id, r.ref_date
        if pd.isna(cid) or pd.isna(rd): continue
        h_cust = cache.get(cid)
        if h_cust is None:
            h_cust = hist_df[hist_df["cust_id"] == cid]
            cache[cid] = h_cust
        f = aggregate_for_one(h_cust, rd); f["cust_id"]=cid; rows.append(f)
        if progress_every and (i % progress_every == 0):
            print(f"[features] {i}/{len(keys_df)}")
    return pd.DataFrame(rows)

# === 4) METRİKLER ===
def gini_from_auc(y_true, y_score):
    return 2 * roc_auc_score(y_true, y_score) - 1

def recall_at_k(y_true, y_score, k=0.10):
    y_true = np.array(y_true)
    topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]
    return float(y_true[idx].sum()) / float(y_true.sum() + 1e-9)

def lift_at_k(y_true, y_score, k=0.10):
    y_true = np.array(y_true)
    topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]
    prec=float(y_true[idx].mean()); prev=float(y_true.mean())
    return prec/max(prev,1e-9)

def competition_score(y_true, y_score):
    g=gini_from_auc(y_true,y_score); r=recall_at_k(y_true,y_score); l=lift_at_k(y_true,y_score)
    return {"Gini":round(g,4),"Recall@10%":round(r,4),"Lift@10%":round(l,4),"CompetitionScore":round(0.4*g+0.3*r+0.3*l,4)}

# === 5) MINI KEY SEÇİMİ ===
def pick_keys_for_mini(ref_df, sample_n=SAMPLE_N, focus_recent=FOCUS_RECENT, recent_frac=RECENT_FRAC):
    keys = ref_df[["cust_id","ref_date","churn"]].copy().sort_values("ref_date").reset_index(drop=True)
    if focus_recent:
        q = keys["ref_date"].quantile(1.0 - recent_frac)
        keys_recent = keys[keys["ref_date"] >= q].reset_index(drop=True)
        if len(keys_recent) > sample_n:
            idx = np.linspace(0, len(keys_recent)-1, sample_n).round().astype(int)
            keys = keys_recent.iloc[idx].reset_index(drop=True)
        else:
            keys = keys_recent
    else:
        if len(keys) > sample_n:
            idx = np.linspace(0, len(keys)-1, sample_n).round().astype(int)
            keys = keys.iloc[idx].reset_index(drop=True)
    return keys

# === 6) Takvim (sızıntısız) ===
def add_calendar_feats(df):
    df["ref_month"]   = df["ref_date"].dt.month.astype(int)
    df["ref_quarter"] = df["ref_date"].dt.quarter.astype(int)
    df["ref_month_sin"] = np.sin(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["ref_month_cos"] = np.cos(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["is_year_start_end"] = ((df["ref_month"]==12) | (df["ref_month"]==1)).astype("Int8")
    return df

# === 7) MODEL EĞİTİMİ (base + opsiyonel pruning) ===
def train_once(X_train, y_train, X_valid, y_valid, cols):
    clf = LGBMClassifier(
        objective="binary",
        n_estimators=1000,
        learning_rate=0.03,
        max_depth=8,
        num_leaves=128,
        subsample=0.90,
        subsample_freq=1,
        colsample_bytree=0.85,
        min_child_samples=25,
        reg_alpha=0.10,
        reg_lambda=1.00,
        class_weight="balanced",
        random_state=SEED
    )
    clf.fit(
        X_train[cols], y_train,
        eval_set=[(X_valid[cols], y_valid)],
        eval_metric='auc',
        callbacks=[early_stopping(stopping_rounds=60), log_evaluation(period=50)]
    )
    proba = clf.predict_proba(X_valid[cols])[:,1]
    metrics = competition_score(y_valid, proba)
    return clf, metrics

def train_eval_mini():
    keys = pick_keys_for_mini(ref, SAMPLE_N, FOCUS_RECENT, RECENT_FRAC)
    print(f"🎯 MINI | keys: {len(keys)}")
    feats = build_features_for_keys(hist, keys[["cust_id","ref_date"]], progress_every=2000)
    X_full = keys.merge(feats, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
    X_full = add_calendar_feats(X_full)
    print("✅ Eğitim seti hazır:", X_full.shape)

    # encode
    num_cols = X_full.select_dtypes(include="number").columns
    cat_cols = [c for c in X_full.columns if c not in num_cols and c != "ref_date"]
    X_full[num_cols] = X_full[num_cols].fillna(0)
    for c in cat_cols: X_full[c] = X_full[c].fillna("Unknown")
    X_full_enc = pd.get_dummies(X_full.drop(columns=["ref_date"]), columns=cat_cols, drop_first=True)
    X_full_enc.columns = X_full_enc.columns.str.replace(r"\s+", "_", regex=True)

    # low-variance drop
    if DO_LOWVAR_DROP:
        nunique = X_full_enc.nunique(dropna=False)
        lowvar_cols = nunique[nunique <= 1].index.tolist()
        if lowvar_cols:
            print(f"🧹 Low-variance drop: {len(lowvar_cols)} kolon")
            X_full_enc = X_full_enc.drop(columns=lowvar_cols)

    # split
    cut = X_full["ref_date"].quantile(0.80)
    train_mask = X_full["ref_date"] < cut
    valid_mask = ~train_mask
    TARGET="churn"; PROTECTED=["cust_id", TARGET]
    feat_cols = [c for c in X_full_enc.columns if c not in PROTECTED]

    X_train = X_full_enc.loc[train_mask, feat_cols]
    y_train = X_full.loc[train_mask, TARGET].astype(int)
    X_valid = X_full_enc.loc[valid_mask, feat_cols]
    y_valid = X_full.loc[valid_mask, TARGET].astype(int)
    print("✅ Split | Train:", X_train.shape, "| Valid:", X_valid.shape)

    # base training
    clf_base, met_base = train_once(X_train, y_train, X_valid, y_valid, feat_cols)
    print("\n📊 BASE:", met_base)

    best_clf, best_cols, best_met = clf_base, feat_cols, met_base

    # pruning (gain tabanlı alt %15)
    if DO_PRUNE:
        try:
            imp = pd.DataFrame({
                "feature": clf_base.feature_name_,
                "importance": clf_base.feature_importances_
            }).sort_values("importance", ascending=True)
            cut_idx = int(len(imp) * PRUNE_FRAC)
            drop_cols = imp.iloc[:cut_idx]["feature"].tolist()
            keep_cols = [c for c in feat_cols if c not in drop_cols]

            clf_p, met_p = train_once(X_train, y_train, X_valid, y_valid, keep_cols)
            print("\n🏁 PRUNED:", met_p, f"(dropped {len(drop_cols)})")

            if met_p["CompetitionScore"] >= met_base["CompetitionScore"]:
                best_clf, best_cols, best_met = clf_p, keep_cols, met_p
            else:
                best_clf, best_cols, best_met = clf_base, feat_cols, met_base
        except Exception as e:
            print("⚠️ Pruning atlandı:", e)

    # özet ve top-20 önem
    print("\n✅ Seçilen skor:", best_met)
    try:
        imp_top = pd.DataFrame({
            "feature": best_clf.feature_name_,
            "importance": best_clf.feature_importances_
        }).sort_values("importance", ascending=False).head(20)
        print("\n🏅 En önemli 20 değişken:")
        print(imp_top)
    except Exception:
        pass

    return best_clf, best_cols, best_met

# === 8) ÇALIŞTIR (Mini Only) ===
if __name__ == "__main__":
    _ = train_eval_mini()
```

```python
# ============================================================
# 🧠 ING Datathon — MINI (LB-1.07 tuned + 3-seed ensemble)
#   - Recent-focused sampling: SAMPLE_N=12000, RECENT_FRAC=0.55
#   - Güçlü feature'lar: last_active / zero_months / HHI / CV / EWM (span6)
#   - Pruning KAPALI (yalnızca açıkça istenirse devreye al)
#   - class_weight="balanced" (winning ayar)
#   - 3-seed ensemble (rank stabilitesi için)
#   - Submission YOK (mini validasyon)
# ============================================================

import os, glob, gc
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier, early_stopping, log_evaluation
from sklearn.metrics import roc_auc_score

pd.set_option("display.width", 160)
pd.set_option("display.max_columns", 220)

# ====== 1) PARAMS ======
SEED = 42
SAMPLE_N     = 12000
FOCUS_RECENT = True
RECENT_FRAC  = 0.55

DO_LOWVAR_DROP = True
DO_PRUNE       = False     # 👈 LB'de oynatmasın diye kapalı tutuyoruz
PRUNE_FRAC     = 0.15      # (açarsan alt %15'i dener)

USE_ENSEMBLE   = True
SEEDS = [(42, 0.85, 0.90), (7, 0.90, 0.85), (2027, 0.80, 0.95)]  # (seed, colsample, subsample)

# ====== 2) DOSYA ======
def find_file(fname: str, search_roots=(".",)):
    fname = fname.lower()
    for root in search_roots:
        for path in glob.iglob(os.path.join(root, "**", "*"), recursive=True):
            if os.path.isfile(path) and os.path.basename(path).lower() == fname:
                return os.path.abspath(path)
    return None

def resolve_path(default_path: str, fallback_name: str):
    if os.path.exists(default_path): return default_path
    p = find_file(fallback_name, search_roots=(".", os.getcwd()))
    if p:
        print(f"[info] Bulundu: {fallback_name} -> {p}")
        return p
    raise FileNotFoundError(f"Dosya bulunamadı: {fallback_name}")

PATH_HISTORY  = resolve_path("/mnt/data/customer_history.csv", "customer_history.csv")
PATH_CUSTOMERS= resolve_path("/mnt/data/customers.csv", "customers.csv")
PATH_REF      = resolve_path("/mnt/data/referance_data.csv", "referance_data.csv")

hist = pd.read_csv(PATH_HISTORY, parse_dates=["date"], low_memory=False)
cust = pd.read_csv(PATH_CUSTOMERS, low_memory=False)
ref  = pd.read_csv(PATH_REF, parse_dates=["ref_date"], low_memory=False)

for df in (hist, cust, ref):
    if "cust_id" in df.columns:
        df["cust_id"] = pd.to_numeric(df["cust_id"], errors="coerce").astype("Int64")
if "churn" in ref.columns:
    ref["churn"] = pd.to_numeric(ref["churn"], errors="coerce").fillna(0).astype("Int8")
cust.columns = cust.columns.str.strip()
hist.columns = hist.columns.str.strip()

print("✅ Dosyalar yüklendi: ",
      "\n  hist:", hist.shape,
      "| cust:", cust.shape,
      "| ref:", ref.shape)

# ====== 3) FEATURING ======
HIST_COLS = [
    "cust_id","date",
    "mobile_eft_all_cnt","active_product_category_nbr",
    "mobile_eft_all_amt","cc_transaction_all_amt","cc_transaction_all_cnt"
]
hist = hist[[c for c in HIST_COLS if c in hist.columns]].sort_values(["cust_id","date"]).reset_index(drop=True)

WINDOWS = [1,3,6,12]
BASE_NUM_COLS = [c for c in HIST_COLS if c not in ["cust_id","date"]]

def _agg_safe(s, fn):
    s = pd.to_numeric(s, errors="coerce"); s = s[np.isfinite(s)]
    if s.size == 0: return 0.0
    if fn=="sum":  return float(s.sum())
    if fn=="mean": return float(s.mean())
    if fn=="std":  return float(s.std(ddof=0))
    if fn=="max":  return float(s.max())
    return 0.0

def _months_active(frame: pd.DataFrame) -> int:
    if frame.empty: return 0
    cond = (pd.to_numeric(frame.get("cc_transaction_all_cnt", 0), errors="coerce").fillna(0) > 0) | \
           (pd.to_numeric(frame.get("mobile_eft_all_cnt", 0), errors="coerce").fillna(0) > 0)
    return int(cond.sum())

def _ewm_last(series: pd.Series, span: int):
    series = pd.to_numeric(series, errors="coerce").fillna(0.0)
    if series.empty: return 0.0
    return float(series.ewm(span=span, adjust=False).mean().iloc[-1])

def aggregate_for_one(h_cust: pd.DataFrame, ref_date: pd.Timestamp):
    feat = {"ref_date": ref_date}
    h = h_cust[h_cust["date"] < ref_date]

    if h.empty:
        for col in BASE_NUM_COLS:
            for w in WINDOWS:
                feat[f"{col}_L{w}M_sum"]=feat[f"{col}_L{w}M_mean"]=feat[f"{col}_L{w}M_std"]=feat[f"{col}_L{w}M_max"]=0.0
        feat["recency_days"] = 9999
        for w in WINDOWS:
            feat[f"months_active_L{w}M"]=0.0
            feat[f"activity_rate_L{w}M"]=0.0
            feat[f"inactivity_ratio_L{w}M"]=1.0
            feat[f"cc_share_amt_L{w}M"]=0.0
            feat[f"eft_share_amt_L{w}M"]=0.0
        feat["cc_amt_ewm_span6"]=feat["eft_amt_ewm_span6"]=0.0
        feat["cc_cnt_ewm_span6"]=feat["eft_cnt_ewm_span6"]=0.0
        feat["last_active_days"]=feat["last_cc_days"]=feat["last_eft_days"]=9999
        feat["zero_cc_months_L3"]=3.0; feat["zero_eft_months_L3"]=3.0
        feat["zero_cc_months_L6"]=6.0; feat["zero_eft_months_L6"]=6.0
    else:
        feat["recency_days"] = int((ref_date - h["date"].max()).days)
        for w in WINDOWS:
            start = ref_date - pd.DateOffset(months=w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            for col in BASE_NUM_COLS:
                feat[f"{col}_L{w}M_sum"]  = _agg_safe(hw[col], "sum")
                feat[f"{col}_L{w}M_mean"] = _agg_safe(hw[col], "mean")
                feat[f"{col}_L{w}M_std"]  = _agg_safe(hw[col], "std")
                feat[f"{col}_L{w}M_max"]  = _agg_safe(hw[col], "max")
            ma = float(_months_active(hw))
            feat[f"months_active_L{w}M"]=ma
            feat[f"activity_rate_L{w}M"]=ma/float(w)
            feat[f"inactivity_ratio_L{w}M"]=float(max(w-ma,0))/float(w)
            eps=1e-6
            cc_sum = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0)
            eft_sum= feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0)
            tot = cc_sum + eft_sum
            feat[f"cc_share_amt_L{w}M"]=cc_sum/(tot+eps)
            feat[f"eft_share_amt_L{w}M"]=eft_sum/(tot+eps)

        # EWM (span6)
        h_sorted = h.sort_values("date")
        feat["cc_amt_ewm_span6"] = _ewm_last(h_sorted["cc_transaction_all_amt"], 6)
        feat["eft_amt_ewm_span6"] = _ewm_last(h_sorted["mobile_eft_all_amt"], 6)
        feat["cc_cnt_ewm_span6"]  = _ewm_last(h_sorted["cc_transaction_all_cnt"], 6)
        feat["eft_cnt_ewm_span6"] = _ewm_last(h_sorted["mobile_eft_all_cnt"], 6)

        # last active (gerçek işlem günü)
        def _last_days(mask):
            if mask.any():
                last_dt = h.loc[mask, "date"].max()
                return int((ref_date - last_dt).days)
            return 9999
        cc_cnt = pd.to_numeric(h.get("cc_transaction_all_cnt", 0), errors="coerce").fillna(0)
        eft_cnt= pd.to_numeric(h.get("mobile_eft_all_cnt", 0), errors="coerce").fillna(0)
        any_act = (cc_cnt>0) | (eft_cnt>0)
        feat["last_active_days"]=_last_days(any_act)
        feat["last_cc_days"]=_last_days(cc_cnt>0)
        feat["last_eft_days"]=_last_days(eft_cnt>0)

        # zero months (3/6)
        def _zero_months(last_w):
            start = ref_date - pd.DateOffset(months=last_w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            ccz = (pd.to_numeric(hw.get("cc_transaction_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            eftz= (pd.to_numeric(hw.get("mobile_eft_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            return float(ccz), float(eftz)
        z3_cc,z3_eft=_zero_months(3); z6_cc,z6_eft=_zero_months(6)
        feat["zero_cc_months_L3"]=z3_cc; feat["zero_eft_months_L3"]=z3_eft
        feat["zero_cc_months_L6"]=z6_cc; feat["zero_eft_months_L6"]=z6_eft

    # işlem başına tutar
    eps=1e-6
    for w in WINDOWS:
        feat[f"cc_amt_per_txn_L{w}M"] = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0)/(feat.get(f"cc_transaction_all_cnt_L{w}M_sum",0.0)+eps)
        feat[f"mobile_eft_amt_per_txn_L{w}M"] = feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0)/(feat.get(f"mobile_eft_all_cnt_L{w}M_sum",0.0)+eps)

    # trendler (mean bazlı)
    for short,long in [(1,3),(3,6),(6,12)]:
        for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt","active_product_category_nbr"]:
            ms = feat.get(f"{base}_L{short}M_mean",0.0)
            ml = feat.get(f"{base}_L{long}M_mean",0.0)
            feat[f"{base}_trend_mean_L{short}vL{long}_diff"]=ms-ml
            feat[f"{base}_trend_mean_L{short}vL{long}_ratio"]=ms/(ml+1e-6)

    # HHI & CV
    for w in WINDOWS:
        cc_share  = feat.get(f"cc_share_amt_L{w}M",0.0)
        eft_share = feat.get(f"eft_share_amt_L{w}M",0.0)
        feat[f"channel_hhi_amt_L{w}M"]=cc_share**2 + eft_share**2

    for base in ["cc_transaction_all_amt","mobile_eft_all_amt","cc_transaction_all_cnt","mobile_eft_all_cnt"]:
        for w in [3,6,12]:
            stdv = feat.get(f"{base}_L{w}M_std",0.0)
            mean = feat.get(f"{base}_L{w}M_mean",0.0)
            feat[f"{base}_cv_L{w}M"]= stdv/(mean+eps)

    # recency dönüşüm
    feat["recency_log"]   = float(np.log1p(feat.get("recency_days",9999)))
    feat["recency_score"] = float(np.exp(-feat.get("recency_days",9999)/30.0))

    return feat

def build_features_for_keys(hist_df, keys_df, progress_every=3000):
    rows, cache = [], {}
    for i, r in enumerate(keys_df.itertuples(index=False), 1):
        cid, rd = r.cust_id, r.ref_date
        if pd.isna(cid) or pd.isna(rd): continue
        h_cust = cache.get(cid)
        if h_cust is None:
            h_cust = hist_df[hist_df["cust_id"] == cid]
            cache[cid] = h_cust
        f = aggregate_for_one(h_cust, rd); f["cust_id"]=cid; rows.append(f)
        if progress_every and (i % progress_every == 0):
            print(f"[features] {i}/{len(keys_df)}")
    return pd.DataFrame(rows)

# ====== 4) METRİK ======
def gini_from_auc(y_true, y_score):
    return 2 * roc_auc_score(y_true, y_score) - 1

def recall_at_k(y_true, y_score, k=0.10):
    y_true = np.array(y_true)
    topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]
    return float(y_true[idx].sum()) / float(y_true.sum() + 1e-9)

def lift_at_k(y_true, y_score, k=0.10):
    y_true = np.array(y_true)
    topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]
    prec=float(y_true[idx].mean()); prev=float(y_true.mean())
    return prec/max(prev,1e-9)

def competition_score(y_true, y_score):
    g=gini_from_auc(y_true,y_score); r=recall_at_k(y_true,y_score); l=lift_at_k(y_true,y_score)
    return {"Gini":round(g,4),"Recall@10%":round(r,4),"Lift@10%":round(l,4),"CompetitionScore":round(0.4*g+0.3*r+0.3*l,4)}

# ====== 5) MINI KEYLER + TAKVİM ======
def pick_keys_for_mini(ref_df, sample_n=SAMPLE_N, focus_recent=FOCUS_RECENT, recent_frac=RECENT_FRAC):
    keys = ref_df[["cust_id","ref_date","churn"]].copy().sort_values("ref_date").reset_index(drop=True)
    if focus_recent:
        q = keys["ref_date"].quantile(1.0 - recent_frac)
        keys_recent = keys[keys["ref_date"] >= q].reset_index(drop=True)
        if len(keys_recent) > sample_n:
            idx = np.linspace(0, len(keys_recent)-1, sample_n).round().astype(int)
            keys = keys_recent.iloc[idx].reset_index(drop=True)
        else:
            keys = keys_recent
    else:
        if len(keys) > sample_n:
            idx = np.linspace(0, len(keys)-1, sample_n).round().astype(int)
            keys = keys.iloc[idx].reset_index(drop=True)
    return keys

def add_calendar_feats(df):
    df["ref_month"]   = df["ref_date"].dt.month.astype(int)
    df["ref_quarter"] = df["ref_date"].dt.quarter.astype(int)
    df["ref_month_sin"] = np.sin(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["ref_month_cos"] = np.cos(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["is_year_start_end"] = ((df["ref_month"]==12) | (df["ref_month"]==1)).astype("Int8")
    return df

# ====== 6) TRAIN ======
def train_once(X_train, y_train, X_valid, y_valid, cols, seed=SEED, colfrac=0.85, rowfrac=0.90):
    clf = LGBMClassifier(
        objective="binary",
        n_estimators=1000,
        learning_rate=0.03,
        max_depth=8,
        num_leaves=128,
        subsample=rowfrac,
        subsample_freq=1,
        colsample_bytree=colfrac,
        min_child_samples=25,
        reg_alpha=0.10,
        reg_lambda=1.00,
        class_weight="balanced",
        random_state=seed
    )
    clf.fit(
        X_train[cols], y_train,
        eval_set=[(X_valid[cols], y_valid)],
        eval_metric='auc',
        callbacks=[early_stopping(stopping_rounds=80), log_evaluation(period=50)]
    )
    proba = clf.predict_proba(X_valid[cols])[:,1]
    metrics = competition_score(y_valid, proba)
    return clf, proba, metrics

def run_mini():
    keys = pick_keys_for_mini(ref)
    print(f"🎯 MINI | keys: {len(keys)}")

    feats = build_features_for_keys(hist, keys[["cust_id","ref_date"]], progress_every=2000)
    X_full = keys.merge(feats, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
    X_full = add_calendar_feats(X_full)
    print("✅ Eğitim seti hazır:", X_full.shape)

    # encode
    num_cols = X_full.select_dtypes(include="number").columns
    cat_cols = [c for c in X_full.columns if c not in num_cols and c != "ref_date"]
    X_full[num_cols] = X_full[num_cols].fillna(0)
    for c in cat_cols: X_full[c] = X_full[c].fillna("Unknown")
    X_full_enc = pd.get_dummies(X_full.drop(columns=["ref_date"]), columns=cat_cols, drop_first=True)
    X_full_enc.columns = X_full_enc.columns.str.replace(r"\s+", "_", regex=True)

    # low-variance drop
    if DO_LOWVAR_DROP:
        nunique = X_full_enc.nunique(dropna=False)
        lowvar_cols = nunique[nunique <= 1].index.tolist()
        if lowvar_cols:
            print(f"🧹 Low-variance drop: {len(lowvar_cols)} kolon")
            X_full_enc = X_full_enc.drop(columns=lowvar_cols)

    # split
    cut = X_full["ref_date"].quantile(0.80)
    train_mask = X_full["ref_date"] < cut
    valid_mask = ~train_mask
    TARGET="churn"; PROTECTED=["cust_id", TARGET]
    feat_cols = [c for c in X_full_enc.columns if c not in PROTECTED]

    X_train = X_full_enc.loc[train_mask, feat_cols]
    y_train = X_full.loc[train_mask, TARGET].astype(int).values
    X_valid = X_full_enc.loc[valid_mask, feat_cols]
    y_valid = X_full.loc[valid_mask, TARGET].astype(int).values

    print("✅ Split | Train:", X_train.shape, "| Valid:", X_valid.shape)

    # base
    base_clf, base_proba, base_met = train_once(X_train, y_train, X_valid, y_valid, feat_cols, seed=SEED)
    print("\n📊 BASE:", base_met)

    best_proba = base_proba
    best_clf   = base_clf
    best_cols  = feat_cols
    best_met   = base_met

    # ensemble
    if USE_ENSEMBLE:
        probs = [base_proba]
        clfs  = [base_clf]
        for sd, colf, rowf in SEEDS[1:]:
            clf_i, proba_i, _ = train_once(X_train, y_train, X_valid, y_valid, feat_cols, seed=sd, colfrac=colf, rowfrac=rowf)
            probs.append(proba_i); clfs.append(clf_i)
        ens_proba = np.mean(np.vstack(probs), axis=0)
        ens_met = competition_score(y_valid, ens_proba)
        print("\n🤝 ENSEMBLE:", ens_met)
        if ens_met["CompetitionScore"] >= best_met["CompetitionScore"]:
            best_proba, best_clf, best_met = ens_proba, clfs[0], ens_met  # (importance için ilkini saklıyoruz)

    # optional pruning (kapalı)
    if DO_PRUNE:
        imp = pd.DataFrame({"feature": best_clf.feature_name_, "importance": best_clf.feature_importances_}).sort_values("importance")
        drop_n = int(len(imp)*PRUNE_FRAC)
        drop_cols = set(imp.iloc[:drop_n]["feature"].tolist())
        keep_cols = [c for c in feat_cols if c not in drop_cols]
        pr_clf, pr_proba, pr_met = train_once(X_train, y_train, X_valid, y_valid, keep_cols, seed=SEED)
        print("\n🪓 PRUNED:", pr_met, f"(dropped {drop_n})")
        if pr_met["CompetitionScore"] >= best_met["CompetitionScore"]:
            best_proba, best_clf, best_cols, best_met = pr_proba, pr_clf, keep_cols, pr_met

    # sonuç & önem
    print("\n✅ Seçilen skor:", best_met)
    try:
        imp_top = pd.DataFrame({"feature": X_train.columns, "importance": best_clf.feature_importances_}).sort_values("importance", ascending=False).head(25)
        print("\n🏅 En önemli 25 değişken:")
        print(imp_top.to_string(index=False))
    except Exception:
        pass

    return best_met

# ====== 7) ÇALIŞTIR ======
if __name__ == "__main__":
    _ = run_mini()
```

```python
# ============================
# MINI — back-to-best (0.90+ push)
#  - SAMPLE_N=12000, RECENT_FRAC=0.55
#  - class_weight="balanced" (NO scale_pos_weight, NO sample weights)
#  - Pruning OFF, only low-variance drop
#  - 5-seed ensemble + rank-avg selection
#  - Added features: acceleration, log1p, seasonal interactions
# ============================
import os, glob, gc
import numpy as np, pandas as pd
from lightgbm import LGBMClassifier, early_stopping, log_evaluation
from sklearn.metrics import roc_auc_score

pd.set_option("display.width", 160)
pd.set_option("display.max_columns", 240)

SEED = 42
SAMPLE_N = 12000
RECENT_FRAC = 0.55
DO_LOWVAR_DROP = True
USE_ENSEMBLE = True

# 5-seed çeşitlendirme (hafif farklı col/row fraction)
SEEDS = [
    (42,   0.85, 0.90),
    (7,    0.90, 0.85),
    (2027, 0.80, 0.95),
    (111,  0.88, 0.88),
    (314,  0.92, 0.80),
]

# -------- I/O yardımcıları --------
def find_file(fname, search_roots=(".",)):
    fname = fname.lower()
    for root in search_roots:
        for path in glob.iglob(os.path.join(root, "**", "*"), recursive=True):
            if os.path.isfile(path) and os.path.basename(path).lower() == fname:
                return os.path.abspath(path)
    return None

def resolve_path(default_path, fallback_name):
    if os.path.exists(default_path): return default_path
    p = find_file(fallback_name, search_roots=(".", os.getcwd()))
    if p:
        print(f"[info] Bulundu: {fallback_name} -> {p}")
        return p
    raise FileNotFoundError(f"Dosya yok: {fallback_name}")

PATH_HISTORY   = resolve_path("/mnt/data/customer_history.csv", "customer_history.csv")
PATH_CUSTOMERS = resolve_path("/mnt/data/customers.csv", "customers.csv")
PATH_REF       = resolve_path("/mnt/data/referance_data.csv", "referance_data.csv")

hist = pd.read_csv(PATH_HISTORY, parse_dates=["date"], low_memory=False)
cust = pd.read_csv(PATH_CUSTOMERS, low_memory=False)
ref  = pd.read_csv(PATH_REF, parse_dates=["ref_date"], low_memory=False)

for df in (hist, cust, ref):
    if "cust_id" in df.columns:
        df["cust_id"] = pd.to_numeric(df["cust_id"], errors="coerce").astype("Int64")
if "churn" in ref.columns:
    ref["churn"] = pd.to_numeric(ref["churn"], errors="coerce").fillna(0).astype("Int8")
for d in (hist, cust):
    d.columns = d.columns.str.strip()

print("✅ Dosyalar:", hist.shape, cust.shape, ref.shape)

# -------- Feature çekirdeği --------
HIST_COLS = [
    "cust_id","date",
    "mobile_eft_all_cnt","active_product_category_nbr",
    "mobile_eft_all_amt","cc_transaction_all_amt","cc_transaction_all_cnt"
]
hist = hist[[c for c in HIST_COLS if c in hist.columns]].sort_values(["cust_id","date"]).reset_index(drop=True)

WINDOWS = [1,3,6,12]
BASE_NUM_COLS = [c for c in HIST_COLS if c not in ["cust_id","date"]]

def _agg_safe(s, fn):
    s = pd.to_numeric(s, errors="coerce"); s = s[np.isfinite(s)]
    if s.size == 0: return 0.0
    if fn=="sum":  return float(s.sum())
    if fn=="mean": return float(s.mean())
    if fn=="std":  return float(s.std(ddof=0))
    if fn=="max":  return float(s.max())
    return 0.0

def _months_active(frame):
    if frame.empty: return 0
    cc = pd.to_numeric(frame.get("cc_transaction_all_cnt",0), errors="coerce").fillna(0)
    ef = pd.to_numeric(frame.get("mobile_eft_all_cnt",0), errors="coerce").fillna(0)
    return int(((cc>0)|(ef>0)).sum())

def _ewm_last(series, span):
    series = pd.to_numeric(series, errors="coerce").fillna(0.0)
    if series.empty: return 0.0
    return float(series.ewm(span=span, adjust=False).mean().iloc[-1])

def _safe_ratio(a, b):
    return float(a) / float(b + 1e-6)

def aggregate_for_one(h_cust, ref_date):
    feat = {"ref_date": ref_date}
    h = h_cust[h_cust["date"] < ref_date]

    if h.empty:
        for col in BASE_NUM_COLS:
            for w in WINDOWS:
                feat[f"{col}_L{w}M_sum"]=feat[f"{col}_L{w}M_mean"]=feat[f"{col}_L{w}M_std"]=feat[f"{col}_L{w}M_max"]=0.0
        feat["recency_days"]=9999
        for w in WINDOWS:
            feat[f"months_active_L{w}M"]=0.0
            feat[f"activity_rate_L{w}M"]=0.0
            feat[f"inactivity_ratio_L{w}M"]=1.0
            feat[f"cc_share_amt_L{w}M"]=0.0
            feat[f"eft_share_amt_L{w}M"]=0.0
        feat["cc_amt_ewm_span6"]=feat["eft_amt_ewm_span6"]=0.0
        feat["cc_cnt_ewm_span6"]=feat["eft_cnt_ewm_span6"]=0.0
        feat["last_active_days"]=feat["last_cc_days"]=feat["last_eft_days"]=9999
        feat["zero_cc_months_L3"]=3.0; feat["zero_eft_months_L3"]=3.0
        feat["zero_cc_months_L6"]=6.0; feat["zero_eft_months_L6"]=6.0
    else:
        feat["recency_days"] = int((ref_date - h["date"].max()).days)
        for w in WINDOWS:
            start = ref_date - pd.DateOffset(months=w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            for col in BASE_NUM_COLS:
                feat[f"{col}_L{w}M_sum"]  = _agg_safe(hw[col], "sum")
                feat[f"{col}_L{w}M_mean"] = _agg_safe(hw[col], "mean")
                feat[f"{col}_L{w}M_std"]  = _agg_safe(hw[col], "std")
                feat[f"{col}_L{w}M_max"]  = _agg_safe(hw[col], "max")
            ma = float(_months_active(hw))
            feat[f"months_active_L{w}M"]=ma
            feat[f"activity_rate_L{w}M"]=_safe_ratio(ma, w)
            feat[f"inactivity_ratio_L{w}M"]=_safe_ratio(max(w-ma,0), w)
            cc_sum  = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0)
            eft_sum = feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0)
            tot = cc_sum + eft_sum
            feat[f"cc_share_amt_L{w}M"] = _safe_ratio(cc_sum, tot)
            feat[f"eft_share_amt_L{w}M"] = _safe_ratio(eft_sum, tot)

        h_sorted = h.sort_values("date")
        feat["cc_amt_ewm_span6"] = _ewm_last(h_sorted["cc_transaction_all_amt"], 6)
        feat["eft_amt_ewm_span6"] = _ewm_last(h_sorted["mobile_eft_all_amt"], 6)
        feat["cc_cnt_ewm_span6"]  = _ewm_last(h_sorted["cc_transaction_all_cnt"], 6)
        feat["eft_cnt_ewm_span6"] = _ewm_last(h_sorted["mobile_eft_all_cnt"], 6)

        def _last_days(mask):
            if mask.any():
                last_dt = h.loc[mask, "date"].max()
                return int((ref_date - last_dt).days)
            return 9999
        cc_cnt = pd.to_numeric(h.get("cc_transaction_all_cnt",0), errors="coerce").fillna(0)
        ef_cnt = pd.to_numeric(h.get("mobile_eft_all_cnt",0), errors="coerce").fillna(0)
        any_act = (cc_cnt>0)|(ef_cnt>0)
        feat["last_active_days"]=_last_days(any_act)
        feat["last_cc_days"]=_last_days(cc_cnt>0)
        feat["last_eft_days"]=_last_days(ef_cnt>0)

        def _zero_months(last_w):
            start = ref_date - pd.DateOffset(months=last_w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            ccz  = (pd.to_numeric(hw.get("cc_transaction_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            eftz = (pd.to_numeric(hw.get("mobile_eft_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            return float(ccz), float(eftz)
        z3c,z3e = _zero_months(3); z6c,z6e = _zero_months(6)
        feat["zero_cc_months_L3"]=z3c; feat["zero_eft_months_L3"]=z3e
        feat["zero_cc_months_L6"]=z6c; feat["zero_eft_months_L6"]=z6e

    # txn başına tutar
    for w in WINDOWS:
        feat[f"cc_amt_per_txn_L{w}M"] = _safe_ratio(
            feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0),
            feat.get(f"cc_transaction_all_cnt_L{w}M_sum",0.0)
        )
        feat[f"mobile_eft_amt_per_txn_L{w}M"] = _safe_ratio(
            feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0),
            feat.get(f"mobile_eft_all_cnt_L{w}M_sum",0.0)
        )

    # trendler (mean bazlı) + acceleration
    for s,l in [(1,3),(3,6),(6,12)]:
        for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt","active_product_category_nbr"]:
            ms = feat.get(f"{base}_L{s}M_mean",0.0)
            ml = feat.get(f"{base}_L{l}M_mean",0.0)
            feat[f"{base}_trend_mean_L{s}vL{l}_diff"]=ms-ml
            feat[f"{base}_trend_mean_L{s}vL{l}_ratio"]=_safe_ratio(ms, ml)

    # acceleration: (L1→L3 / L3→L6) oranı (amt ve cnt için)
    for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt"]:
        m1 = feat.get(f"{base}_L1M_mean",0.0);  m3 = feat.get(f"{base}_L3M_mean",0.0);  m6 = feat.get(f"{base}_L6M_mean",0.0)
        r13 = _safe_ratio(m1, m3); r36 = _safe_ratio(m3, m6)
        feat[f"{base}_accel_L1_3_6"] = _safe_ratio(r13, r36)

    # HHI & CV
    for w in WINDOWS:
        cc_share  = feat.get(f"cc_share_amt_L{w}M",0.0)
        eft_share = feat.get(f"eft_share_amt_L{w}M",0.0)
        feat[f"channel_hhi_amt_L{w}M"]=cc_share**2 + eft_share**2

    for base in ["cc_transaction_all_amt","mobile_eft_all_amt","cc_transaction_all_cnt","mobile_eft_all_cnt"]:
        for w in [3,6,12]:
            stdv = feat.get(f"{base}_L{w}M_std",0.0)
            mean = feat.get(f"{base}_L{w}M_mean",0.0)
            feat[f"{base}_cv_L{w}M"] = _safe_ratio(stdv, mean)

    # log1p stabilizasyon (sadece mean/sum üzerinde seçici)
    for base in ["cc_transaction_all_amt","mobile_eft_all_amt","cc_transaction_all_cnt","mobile_eft_all_cnt","active_product_category_nbr"]:
        for w in [1,3,6,12]:
            feat[f"{base}_L{w}M_mean_log1p"] = float(np.log1p(feat.get(f"{base}_L{w}M_mean", 0.0)))
            feat[f"{base}_L{w}M_sum_log1p"]  = float(np.log1p(feat.get(f"{base}_L{w}M_sum", 0.0)))

    # recency dönüşümleri
    feat["recency_log"]   = float(np.log1p(feat.get("recency_days",9999)))
    feat["recency_score"] = float(np.exp(-feat.get("recency_days",9999)/30.0))

    return feat

def build_features_for_keys(hist_df, keys_df, progress_every=2000):
    rows, cache = [], {}
    for i, r in enumerate(keys_df.itertuples(index=False), 1):
        cid, rd = r.cust_id, r.ref_date
        if pd.isna(cid) or pd.isna(rd): continue
        h_cust = cache.get(cid)
        if h_cust is None:
            h_cust = hist_df[hist_df["cust_id"] == cid]
            cache[cid] = h_cust
        f = aggregate_for_one(h_cust, rd); f["cust_id"]=cid; rows.append(f)
        if progress_every and (i % progress_every == 0):
            print(f"[features] {i}/{len(keys_df)}")
    return pd.DataFrame(rows)

# -------- Metrikler --------
def gini_from_auc(y_true, y_score): return 2*roc_auc_score(y_true, y_score)-1

def recall_at_k(y_true, y_score, k=0.10):
    y_true = np.asarray(y_true); topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]; return float(y_true[idx].sum()) / float(y_true.sum() + 1e-9)

def lift_at_k(y_true, y_score, k=0.10):
    y_true = np.asarray(y_true); topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]; prec=float(y_true[idx].mean()); prev=float(y_true.mean())
    return prec/max(prev,1e-9)

def competition_score(y_true, y_score):
    g=gini_from_auc(y_true,y_score); r=recall_at_k(y_true,y_score); l=lift_at_k(y_true,y_score)
    return {"Gini":round(g,4),"Recall@10%":round(r,4),"Lift@10%":round(l,4),"CompetitionScore":round(0.4*g+0.3*r+0.3*l,4)}

# -------- Mini key seçimi & takvim --------
def pick_keys_for_mini(ref_df, sample_n=SAMPLE_N, recent_frac=RECENT_FRAC):
    keys = ref_df[["cust_id","ref_date","churn"]].copy().sort_values("ref_date").reset_index(drop=True)
    q = keys["ref_date"].quantile(1.0 - recent_frac)
    keys_recent = keys[keys["ref_date"] >= q].reset_index(drop=True)
    if len(keys_recent) > sample_n:
        idx = np.linspace(0, len(keys_recent)-1, sample_n).round().astype(int)
        return keys_recent.iloc[idx].reset_index(drop=True)
    return keys_recent

def add_calendar_feats(df):
    df["ref_month"]   = df["ref_date"].dt.month.astype(int)
    df["ref_quarter"] = df["ref_date"].dt.quarter.astype(int)
    df["ref_month_sin"] = np.sin(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["ref_month_cos"] = np.cos(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["is_year_start_end"] = ((df["ref_month"]==12) | (df["ref_month"]==1)).astype("Int8")
    return df

# -------- Model eğitimi --------
def train_once(X_train, y_train, X_valid, y_valid, cols, seed=SEED, colfrac=0.85, rowfrac=0.90):
    clf = LGBMClassifier(
        objective="binary",
        n_estimators=2000,          # biraz daha yüksek + early stopping
        learning_rate=0.025,        # küçük adım
        max_depth=8,
        num_leaves=144,             # 128'den hafif yukarı (çeşitlilik)
        subsample=rowfrac,
        subsample_freq=1,
        colsample_bytree=colfrac,
        min_child_samples=25,
        reg_alpha=0.10,
        reg_lambda=1.10,
        min_split_gain=0.0,
        class_weight="balanced",
        random_state=seed
    )
    clf.fit(
        X_train[cols], y_train,
        eval_set=[(X_valid[cols], y_valid)],
        eval_metric='auc',
        callbacks=[early_stopping(stopping_rounds=120), log_evaluation(period=50)]
    )
    proba = clf.predict_proba(X_valid[cols])[:,1]
    metrics = competition_score(y_valid, proba)
    return clf, proba, metrics

def run_mini():
    keys = pick_keys_for_mini(ref)
    print(f"🎯 MINI | keys: {len(keys)}")

    feats = build_features_for_keys(hist, keys[["cust_id","ref_date"]], progress_every=2000)
    X_full = keys.merge(feats, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
    X_full = add_calendar_feats(X_full)

    # sezonluk etkileşimler (sade ve sızıntısız)
    X_full["season_x_activity_L3"] = X_full["is_year_start_end"].astype(float) * X_full.get("activity_rate_L3M", 0.0)
    X_full["season_x_recency"]     = X_full["is_year_start_end"].astype(float) * X_full.get("recency_score", 0.0)

    print("✅ Eğitim seti hazır:", X_full.shape)

    # encode
    num_cols = X_full.select_dtypes(include="number").columns
    cat_cols = [c for c in X_full.columns if c not in num_cols and c!="ref_date"]
    X_full[num_cols] = X_full[num_cols].fillna(0)
    for c in cat_cols: X_full[c] = X_full[c].fillna("Unknown")
    X_full_enc = pd.get_dummies(X_full.drop(columns=["ref_date"]), columns=cat_cols, drop_first=True)
    X_full_enc.columns = X_full_enc.columns.str.replace(r"\s+", "_", regex=True)

    # low-variance drop
    if DO_LOWVAR_DROP:
        nunique = X_full_enc.nunique(dropna=False)
        lowvar_cols = nunique[nunique <= 1].index.tolist()
        if lowvar_cols:
            print(f"🧹 Low-variance drop: {len(lowvar_cols)} kolon")
            X_full_enc = X_full_enc.drop(columns=lowvar_cols)

    # time-based split (%80 kesim)
    cut = X_full["ref_date"].quantile(0.80)
    train_mask = X_full["ref_date"] < cut
    valid_mask = ~train_mask

    TARGET="churn"; PROTECTED=["cust_id", TARGET]
    feat_cols = [c for c in X_full_enc.columns if c not in PROTECTED]
    X_train = X_full_enc.loc[train_mask, feat_cols]
    y_train = X_full.loc[train_mask, TARGET].astype(int).values
    X_valid = X_full_enc.loc[valid_mask, feat_cols]
    y_valid = X_full.loc[valid_mask, TARGET].astype(int).values
    print("✅ Split | Train:", X_train.shape, "| Valid:", X_valid.shape)

    # base + ensemble
    clfs, probas = [], []
    for sd, colf, rowf in SEEDS:
        clf_i, proba_i, met_i = train_once(X_train, y_train, X_valid, y_valid, feat_cols, seed=sd, colfrac=colf, rowfrac=rowf)
        print(f"[seed {sd}] {met_i}")
        clfs.append(clf_i); probas.append(proba_i)

    # Olasılık ortalaması
    proba_mean = np.mean(np.vstack(probas), axis=0)
    met_mean = competition_score(y_valid, proba_mean)
    print("\n🤝 ENSEMBLE (proba-mean):", met_mean)

    # Rank ortalaması (top-k metriklerinde genelde daha iyi)
    ranks = [pd.Series(p).rank(method="average")/len(p) for p in probas]
    proba_rank = np.mean(np.vstack([r.values for r in ranks]), axis=0)
    met_rank = competition_score(y_valid, proba_rank)
    print("📈 ENSEMBLE (rank-avg):   ", met_rank)

    # En iyisini seç
    if met_rank["CompetitionScore"] >= met_mean["CompetitionScore"]:
        best_met, best_pred = met_rank, proba_rank
        chosen = "rank-avg"
    else:
        best_met, best_pred = met_mean, proba_mean
        chosen = "proba-mean"

    print(f"\n✅ Seçilen yöntem: {chosen} | Skor: {best_met}")

    # önem listesi (ilk modelden)
    try:
        imp_top = pd.DataFrame({"feature": feat_cols, "importance": clfs[0].feature_importances_})\
                    .sort_values("importance", ascending=False).head(25)
        print("\n🏅 En önemli 25 değişken:")
        print(imp_top.to_string(index=False))
    except Exception:
        pass

    return best_met

if __name__ == "__main__":
    _ = run_mini()
```

```python
# =======================================================
# FULL PIPELINE: Base LGBM -> Isotonic -> Monotonic Rerank
# Hedef: CompetitionScore ↑ (Gini + Lift@10)
# =======================================================

import os, gc, warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.isotonic import IsotonicRegression
import lightgbm as lgb

# ---------------- CONFIG (kendine göre düzenle) ----------------
TRAIN_PATH = r"train_features.csv"     # hazır feature'lı train dosyan
TARGET_COL = "target"                  # 0/1 hedef kolonun adı
ID_COL = "customer_id"                 # varsa id; yoksa None yap
HAS_TEST = False                       # test varsa True
TEST_PATH = r"test_features.csv"       # test dosya yolu (HAS_TEST=True ise)
TOP_K = 0.10                           # Lift/Recall@top% (yarışma 10% ise 0.10)
W_GINI = 0.35                          # skor ağırlıkları (yarışmana göre ayarla)
W_LIFT_MINUS1 = 0.65

SEED = 2027
N_FOLDS = 5

# Reranker'a verilecek ek kolonlar (train/test setindeki isimlerle birebir olmalı)
EXTRA_COLS = [
    "active_product_category_nbr_L12M_sum",
    "age",
    "tenure",
    "mobile_eft_all_amt_trend_mean_L1vL3_diff",
    "mobile_eft_all_cnt_trend_mean_L3vL6_diff",
    "cc_transaction_all_amt_cv_L3M",
    "cc_transaction_all_cnt_cv_L3M",
    "mobile_eft_all_amt_trend_mean_L6vL12_ratio",
    "mobile_eft_all_cnt_L12M_sum",
    "cc_transaction_all_cnt_trend_mean_L3vL6_diff",
]

# -------------------- Metrics --------------------
def recall_at_k(y_true, scores, k=0.10):
    y_true = np.asarray(y_true).astype(int)
    scores = np.asarray(scores)
    n = len(scores); top = int(np.ceil(n*k))
    idx = np.argsort(-scores)[:top]
    return y_true[idx].sum() / (y_true.sum() + 1e-12)

def lift_at_k(y_true, scores, k=0.10):
    r = recall_at_k(y_true, scores, k)
    return r / k

def gini_from_scores(y_true, scores):
    auc = roc_auc_score(y_true, scores)
    return 2*auc - 1

# Yarışmanın formülü farklıysa burada düzenle
def competition_score(y_true, scores, k=0.10, w_gini=0.35, w_liftm1=0.65):
    g = gini_from_scores(y_true, scores)
    L = lift_at_k(y_true, scores, k) - 1.0
    return w_gini*g + w_liftm1*L

def summarize(tag, y, s, k=0.10, w_gini=0.35, w_liftm1=0.65):
    g = gini_from_scores(y, s)
    r = recall_at_k(y, s, k)
    L = r/k
    comp = competition_score(y, s, k, w_gini, w_liftm1)
    print(f"[{tag}]  Gini={g:.4f} | Recall@{int(k*100)}%={r:.4f} | Lift@{int(k*100)}%={L:.4f} | Comp={comp:.4f}")
    return dict(Gini=g, Recall=r, Lift=L, CompetitionScore=comp)

# -------------------- Helpers --------------------
def _logit(p):
    p = np.clip(np.asarray(p), 1e-6, 1-1e-6)
    return np.log(p/(1-p))

def _rank01(s):
    s = np.asarray(s)
    r = pd.Series(s).rank(method="average", ascending=False).to_numpy()
    return (r-1) / max(len(s)-1, 1)

def build_meta_features(base_scores, X=None, take_cols=None):
    z = pd.DataFrame({
        "base": base_scores,
        "base_logit": _logit(base_scores),
        "base_rank01": _rank01(base_scores),
        "base_decile": pd.qcut(base_scores, 10, labels=False, duplicates="drop")
    })
    if X is not None and take_cols:
        keep = [c for c in take_cols if c in X.columns]
        if keep:
            z = pd.concat([z.reset_index(drop=True), X[keep].reset_index(drop=True)], axis=1)
    return z

# -------------------- Base LGBM CV --------------------
def cv_lightgbm_base(X, y, n_folds=5, seed=2027):
    params = dict(
        objective="binary",
        metric="auc",
        learning_rate=0.03,
        num_leaves=128,
        feature_fraction=0.85,
        bagging_fraction=0.9,
        bagging_freq=1,
        min_data_in_leaf=50,
        min_data_in_bin=64,
        lambda_l1=0.0,
        lambda_l2=2.0,
        max_depth=-1,
        verbosity=-1,
        force_col_wise=True,
    )
    skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=seed)
    oof = np.zeros(len(X), dtype=float)
    models = []
    for fold, (tr, va) in enumerate(skf.split(X, y)):
        dtr = lgb.Dataset(X.iloc[tr], label=y.iloc[tr], free_raw_data=False)
        dva = lgb.Dataset(X.iloc[va], label=y.iloc[va], free_raw_data=False)
        mdl = lgb.train(
            params, dtr,
            num_boost_round=4000,
            valid_sets=[dtr, dva],
            valid_names=["tr", "va"],
            early_stopping_rounds=200,
            verbose_eval=False
        )
        oof[va] = mdl.predict(X.iloc[va], num_iteration=mdl.best_iteration)
        models.append(mdl)
        print(f"  fold{fold} AUC={roc_auc_score(y.iloc[va], oof[va]):.4f}")
    return oof, models

# -------------------- Isotonic (OOF) --------------------
def isotonic_top_weight_OOF(base_oof, y,
                            pos_boost=1.3,
                            blend_grid=np.linspace(0, 0.8, 17), k=0.10,
                            w_gini=0.35, w_liftm1=0.65, base_test=None):
    base = np.asarray(base_oof)
    y = np.asarray(y).astype(int)
    r01 = _rank01(base)
    w = 1.0/np.sqrt(1e-3 + r01)                   # tepeye ağırlık
    w *= np.where(y==1, pos_boost, 1.0)           # pozitifleri hafifçe güçlendir

    iso = IsotonicRegression(y_min=0.0, y_max=1.0, increasing=True, out_of_bounds="clip")
    iso.fit(base, y, sample_weight=w)
    oof_iso = iso.transform(base)
    summarize("ISO-ONLY", y, oof_iso, k, w_gini, w_liftm1)

    best = (-9, None, None)
    for a in blend_grid:                           # s = (1-a)*base + a*iso
        s = (1-a)*base + a*oof_iso
        comp = competition_score(y, s, k, w_gini, w_liftm1)
        if comp > best[0]:
            best = (comp, a, s)
    print(f"[ISO-BLEND] best alpha={best[1]:.2f} | Comp={best[0]:.4f}")
    summarize("ISO-BLEND", y, best[2], k, w_gini, w_liftm1)

    test_pred = None
    if base_test is not None:
        test_pred = (1-best[1])*np.asarray(base_test) + best[1]*iso.transform(np.asarray(base_test))

    return dict(oof=best[2], alpha=best[1], iso_model=iso, iso_oof=oof_iso, test=test_pred)

# -------------------- Monotonic Reranker (OOF) --------------------
def monotonic_top_weight_rerank_OOF(base_oof, X, y, extra_cols=None,
                                    n_splits=5, seed=2027,
                                    top_focus_gamma=1.2, pos_boost=1.5,
                                    params=None, num_boost_round=4000, esr=200,
                                    k=0.10, w_gini=0.35, w_liftm1=0.65,
                                    blend_grid=np.linspace(0.0, 0.9, 19),
                                    base_test=None, X_test=None):
    y = pd.Series(y).astype(int)
    Z = build_meta_features(base_oof, X, take_cols=extra_cols)

    if params is None:
        params = dict(
            objective="binary",
            metric="auc",
            learning_rate=0.04,
            num_leaves=127,
            min_data_in_leaf=40,
            feature_fraction=0.85,
            bagging_fraction=0.9,
            bagging_freq=1,
            verbosity=-1,
            force_col_wise=True,
        )

    # base ve base_logit artan zorunlu (monotone)
    mono = []
    for c in Z.columns:
        if c in ["base", "base_logit"]: mono.append(1)
        else: mono.append(0)
    params["monotone_constraints"] = mono

    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=seed)
    oof = np.zeros(len(Z))
    models = []

    base_oof = np.asarray(base_oof)
    r01_all = _rank01(base_oof)

    for fold,(tr,va) in enumerate(skf.split(Z, y)):
        Z_tr, Z_va = Z.iloc[tr], Z.iloc[va]
        y_tr, y_va = y.iloc[tr], y.iloc[va]

        w_tr = np.ones(len(tr), dtype=float)
        w_tr *= 1.0/np.power(1e-3 + r01_all[tr], top_focus_gamma)   # top focus
        w_tr *= np.where(y_tr.values==1, pos_boost, 1.0)            # pozitif boost

        dtr = lgb.Dataset(Z_tr, label=y_tr, weight=w_tr, free_raw_data=False)
        dva = lgb.Dataset(Z_va, label=y_va, free_raw_data=False)
        mdl = lgb.train(params, dtr, num_boost_round=num_boost_round,
                        valid_sets=[dtr,dva], valid_names=["tr","va"],
                        early_stopping_rounds=esr, verbose_eval=False)
        oof[va] = mdl.predict(Z_va, num_iteration=mdl.best_iteration)
        models.append(mdl)

    summarize("RERANK_OOF_RAW", y.values, oof, k, w_gini, w_liftm1)

    # blend (baz ile ne kadar karışacak?)
    best = (-9, None, None)
    for a in blend_grid:
        s = (1-a)*base_oof + a*oof
        comp = competition_score(y.values, s, k, w_gini, w_liftm1)
        if comp > best[0]:
            best = (comp, a, s)
    print(f"[RERANK-BLEND] best alpha={best[1]:.2f} | Comp={best[0]:.4f}")
    summarize("RERANK-BLEND", y.values, best[2], k, w_gini, w_liftm1)

    # test tahmini (varsa)
    test_pred = None
    if (base_test is not None) and (X_test is not None):
        Zt = build_meta_features(base_test, X_test, take_cols=extra_cols)
        fold_preds = np.zeros(len(Zt))
        for mdl in models:
            fold_preds += mdl.predict(Zt, num_iteration=mdl.best_iteration)
        fold_preds /= len(models)
        test_pred = (1-best[1])*np.asarray(base_test) + best[1]*fold_preds

    return dict(oof=best[2], alpha=best[1], raw_oof=oof, models=models, test=test_pred)

# -------------------- Orchestrator --------------------
def top_focus_boost(base_oof, X, y,
                    base_test=None, X_test=None,
                    extra_cols=None,
                    k=0.10,
                    w_gini=0.35, w_liftm1=0.65,
                    seed=2027):
    # 1) Isotonic
    iso = isotonic_top_weight_OOF(
        base_oof=base_oof, y=y, pos_boost=1.3,
        k=k, w_gini=w_gini, w_liftm1=w_liftm1, base_test=base_test
    )
    # 2) Monotonic rerank
    rer = monotonic_top_weight_rerank_OOF(
        base_oof=base_oof, X=X, y=y, extra_cols=extra_cols,
        n_splits=5, seed=seed,
        top_focus_gamma=1.2, pos_boost=1.5,
        k=k, w_gini=w_gini, w_liftm1=w_liftm1,
        base_test=base_test, X_test=X_test
    )

    comp_iso = competition_score(y, iso["oof"], k, w_gini, w_liftm1)
    comp_rer = competition_score(y, rer["oof"], k, w_gini, w_liftm1)
    if comp_rer >= comp_iso:
        print(f"\n[FINAL] seçilen: RERANK-BLEND (Comp={comp_rer:.4f} ≥ {comp_iso:.4f})")
        return dict(method="RERANK", oof=rer["oof"], test=rer["test"],
                    alpha=rer["alpha"], details=dict(iso=iso, rer=rer))
    else:
        print(f"\n[FINAL] seçilen: ISO-BLEND (Comp={comp_iso:.4f} > {comp_rer:.4f})")
        return dict(method="ISO", oof=iso["oof"], test=iso["test"],
                    alpha=iso["alpha"], details=dict(iso=iso, rer=rer))

# -------------------- Data utils --------------------
def load_data(train_path):
    assert os.path.exists(train_path), f"Dosya yok: {train_path}"
    df = pd.read_csv(train_path)
    # hedefi bul
    if TARGET_COL not in df.columns:
        # birkaç yaygın isim için fallback
        cands = [c for c in ["label","y","target","TARGET","is_churn","default"] if c in df.columns]
        assert cands, f"Hedef kolon bulunamadı. Lütfen TARGET_COL değişkenini doğru ayarla."
        target = cands[0]
    else:
        target = TARGET_COL
    # ID varsa tut
    id_col = ID_COL if (ID_COL and ID_COL in df.columns) else None

    # sadece sayısal feature'lar
    drop_cols = [target] + ([id_col] if id_col else [])
    feats = [c for c in df.columns if c not in drop_cols and pd.api.types.is_numeric_dtype(df[c])]
    X = df[feats].copy()
    y = df[target].astype(int).copy()

    # eksik doldur + düşük varyans temizliği
    X = X.replace([np.inf, -np.inf], np.nan)
    X = X.fillna(X.median())
    nunique = X.nunique()
    lowvar = nunique[nunique <= 1].index.tolist()
    if lowvar:
        X = X.drop(columns=lowvar)
        print(f"Low-variance drop: {len(lowvar)} kolon")
    X = X.astype(np.float32)

    return X, y, id_col

def load_test(test_path, feats_to_keep, id_col):
    assert os.path.exists(test_path), f"Dosya yok: {test_path}"
    df = pd.read_csv(test_path)
    if id_col and id_col not in df.columns:
        id_col = None
    # sadece train'de kalan feature'ları seç
    feats = [c for c in feats_to_keep if c in df.columns and pd.api.types.is_numeric_dtype(df[c])]
    Xt = df[feats].copy()
    Xt = Xt.replace([np.inf, -np.inf], np.nan)
    Xt = Xt.fillna(Xt.median()).astype(np.float32)
    return Xt, (df[id_col] if id_col else None)

# -------------------- MAIN --------------------
if __name__ == "__main__":
    np.random.seed(SEED)

    print(">> Train yükleniyor...")
    X, y, id_col = load_data(TRAIN_PATH)
    print(f"Train shape: {X.shape}")

    # ---- Base OOF
    print("\n>> Base LGBM CV başlıyor...")
    base_oof, base_models = cv_lightgbm_base(X, y, n_folds=N_FOLDS, seed=SEED)
    summarize("BASE_OOF", y, base_oof, k=TOP_K, w_gini=W_GINI, w_liftm1=W_LIFT_MINUS1)

    # ---- Test (varsa) base pred
    base_test = None; X_test = None; test_ids = None
    if HAS_TEST:
        print("\n>> Test yükleniyor...")
        X_test, test_ids = load_test(TEST_PATH, X.columns.tolist(), id_col)
        # CV ortalaması
        base_test = np.zeros(len(X_test))
        for m in base_models:
            base_test += m.predict(X_test, num_iteration=m.best_iteration)
        base_test /= len(base_models)

    # ---- Top-Focus Boost (Isotonic + Monotonic Rerank, auto-select)
    print("\n>> Top-Focus Boost (Isotonic + Rerank)...")
    res = top_focus_boost(
        base_oof=base_oof, X=X, y=y,
        base_test=base_test, X_test=X_test,
        extra_cols=EXTRA_COLS,
        k=TOP_K, w_gini=W_GINI, w_liftm1=W_LIFT_MINUS1,
        seed=SEED
    )

    final_oof = res["oof"]
    summarize("FINAL_OOF", y, final_oof, k=TOP_K, w_gini=W_GINI, w_liftm1=W_LIFT_MINUS1)

    # ---- Kaydet
    pd.Series(final_oof, name="oof_final").to_csv("oof_final.csv", index=False)
    print('OOF kaydedildi: oof_final.csv')

    if HAS_TEST and res.get("test") is not None:
        out = pd.DataFrame({"pred": res["test"]})
        if test_ids is not None:
            out.insert(0, id_col, test_ids.values)
        out.to_csv("pred_final.csv", index=False)
        print('Test tahmin kaydedildi: pred_final.csv')
```

```python
def find_file(fname: str, search_roots=(".",)):
    fname = fname.lower()
    for root in search_roots:
        for path in glob.iglob(os.path.join(root, "**", "*"), recursive=True):
            if os.path.isfile(path) and os.path.basename(path).lower() == fname:
                return os.path.abspath(path)
    return None

def resolve_path(default_path: str, fallback_name: str):
    if os.path.exists(default_path):
        return default_path
    p = find_file(fallback_name, search_roots=(".", os.getcwd()))
    if p:
        print(f"[info] Bulundu: {fallback_name} -> {p}")
        return p
    raise FileNotFoundError(f"Dosya bulunamadı: {fallback_name}")
```

```python
HIST_COLS = [
    "cust_id","date",
    "mobile_eft_all_cnt","active_product_category_nbr",
    "mobile_eft_all_amt","cc_transaction_all_amt","cc_transaction_all_cnt"
]
hist = hist[[c for c in HIST_COLS if c in hist.columns]].sort_values(["cust_id","date"]).reset_index(drop=True)
WINDOWS = [1,3,6,12]
BASE_NUM_COLS = [c for c in HIST_COLS if c not in ["cust_id","date"]]

def _agg_safe(s, fn):
    s = pd.to_numeric(s, errors="coerce"); s = s[np.isfinite(s)]
    if s.size == 0: return 0.0
    if fn=="sum":  return float(s.sum())
    if fn=="mean": return float(s.mean())
    if fn=="std":  return float(s.std(ddof=0))
    if fn=="max":  return float(s.max())
    return 0.0

def _months_active(frame):
    if frame.empty: return 0
    cond = (pd.to_numeric(frame.get("cc_transaction_all_cnt", 0), errors="coerce").fillna(0) > 0) | \
           (pd.to_numeric(frame.get("mobile_eft_all_cnt", 0), errors="coerce").fillna(0) > 0)
    return int(cond.sum())

def aggregate_for_one(h_cust, ref_date):
    feat = {"ref_date": ref_date}
    h = h_cust[h_cust["date"] < ref_date]
    if h.empty:
        for col in BASE_NUM_COLS:
            for w in WINDOWS:
                for fn in ["sum","mean","std","max"]:
                    feat[f"{col}_L{w}M_{fn}"] = 0.0
        feat["recency_days"] = 9999
        return feat

    feat["recency_days"] = int((ref_date - h["date"].max()).days)
    for w in WINDOWS:
        start = ref_date - pd.DateOffset(months=w)
        hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
        for col in BASE_NUM_COLS:
            feat[f"{col}_L{w}M_sum"]  = _agg_safe(hw[col], "sum")
            feat[f"{col}_L{w}M_mean"] = _agg_safe(hw[col], "mean")
            feat[f"{col}_L{w}M_std"]  = _agg_safe(hw[col], "std")
            feat[f"{col}_L{w}M_max"]  = _agg_safe(hw[col], "max")
    return feat

def build_features_for_keys(hist_df, keys_df, progress_every=5000):
    rows, cache = [], {}
    for i, r in enumerate(keys_df.itertuples(index=False), 1):
        cid, rd = r.cust_id, r.ref_date
        if pd.isna(cid) or pd.isna(rd): continue
        h_cust = cache.get(cid)
        if h_cust is None:
            h_cust = hist_df[hist_df["cust_id"] == cid]
            cache[cid] = h_cust
        f = aggregate_for_one(h_cust, rd); f["cust_id"]=cid; rows.append(f)
        if progress_every and (i % progress_every == 0):
            print(f"[features] {i}/{len(keys_df)}")
    return pd.DataFrame(rows)
```

```python
def add_calendar_feats(df):
    df["ref_month"]   = df["ref_date"].dt.month.astype(int)
    df["ref_quarter"] = df["ref_date"].dt.quarter.astype(int)
    return df
```

```python
# ============================================================
# ING Datathon — FULL TRAIN + SUBMISSION (LGBM, Native Categorical)
# - Feature'lar: L{1,3,6,12} sum/mean/std/max, months_active/activity_rate,
#   per-txn, trend diff/ratio (1v3,3v6,6v12), HHI, CV, recency(log/score),
#   EWM spans {3,6,12}, last_active/zero_months {3,6,12}, lineer trend slope {6,12},
#   mikro etkileşim: age*tenure, log1p(tenure)
# - Tuning: Zaman dilimi CV + Optuna (yoksa mikro grid)
# - Tek model (ansambl yok). Native categorical split (one-hot yok).
# - Final: full-train, submission.csv
# ============================================================

import os, glob, gc, math, warnings
import numpy as np
import pandas as pd
from datetime import datetime
from lightgbm import LGBMClassifier, early_stopping, log_evaluation
from sklearn.metrics import roc_auc_score

warnings.filterwarnings("ignore")
pd.set_option("display.width", 180)
pd.set_option("display.max_columns", 240)

# --------------------
# 0) PARAMS
# --------------------
SEED = 42
DO_TUNE           = True      # Optuna ile zaman-bilinçli tuning (varsa)
TUNE_SAMPLE_N     = 50000     # tuning keys sayısı (makineye göre 30k–80k arası)
TUNE_RECENT_FRAC  = 0.60      # tuningde güncele hafif vurgu
VALID_QS          = [0.70, 0.80, 0.90]  # rolling zaman dilimi validasyon kesitleri
EARLY_STOP_ROUNDS = 80
BASE_N_EST        = 5000      # early stopping ile kontrol edilir
SUBMISSION_PATH   = "submission.csv"

# Final eğitim için filtre (genelde TÜM veriyi kullan)
TRAIN_RECENT_FRAC_FULL = 1.00  # 1.0 = tüm ref; 0.60 yaparsan güncele ağırlık verirsin.

# --------------------
# 1) IO Helpers
# --------------------
def find_file(fname: str, search_roots=(".",)):
    fname = fname.lower()
    for root in search_roots:
        for path in glob.iglob(os.path.join(root, "**", "*"), recursive=True):
            if os.path.isfile(path) and os.path.basename(path).lower() == fname:
                return os.path.abspath(path)
    return None

def resolve_path(default_path: str, fallback_name: str):
    if os.path.exists(default_path): return default_path
    p = find_file(fallback_name, search_roots=(".", os.getcwd()))
    if p:
        print(f"[info] Bulundu: {fallback_name} -> {p}")
        return p
    raise FileNotFoundError(f"Dosya bulunamadı: {fallback_name}")

PATH_HISTORY   = resolve_path("/mnt/data/customer_history.csv", "customer_history.csv")
PATH_CUSTOMERS = resolve_path("/mnt/data/customers.csv",       "customers.csv")
PATH_REF       = resolve_path("/mnt/data/referance_data.csv",  "referance_data.csv")

SUB_KEYS = None
for alt in ["submission_keys.csv","application_keys.csv","test_keys.csv"]:
    p = find_file(alt, search_roots=(".", os.getcwd()))
    if p:
        SUB_KEYS = p
        print(f"[info] Submission keys bulundu: {alt} -> {p}")
        break

hist = pd.read_csv(PATH_HISTORY, parse_dates=["date"], low_memory=False)
cust = pd.read_csv(PATH_CUSTOMERS, low_memory=False)
ref  = pd.read_csv(PATH_REF, parse_dates=["ref_date"], low_memory=False)

for df in (hist, cust, ref):
    if "cust_id" in df.columns:
        df["cust_id"] = pd.to_numeric(df["cust_id"], errors="coerce").astype("Int64")
if "churn" in ref.columns:
    ref["churn"] = pd.to_numeric(ref["churn"], errors="coerce").fillna(0).astype("Int8")

cust.columns = cust.columns.str.strip()
hist.columns = hist.columns.str.strip()

print("✅ Dosyalar:",
      "\n  hist:", hist.shape,
      "| cust:", cust.shape,
      "| ref:", ref.shape)

# --------------------
# 2) Feature Engineering
# --------------------
HIST_COLS = ["cust_id","date",
             "mobile_eft_all_cnt","active_product_category_nbr",
             "mobile_eft_all_amt","cc_transaction_all_amt","cc_transaction_all_cnt"]
hist = hist[[c for c in HIST_COLS if c in hist.columns]].sort_values(["cust_id","date"]).reset_index(drop=True)

WINDOWS = [1,3,6,12]
BASE_NUM_COLS = [c for c in HIST_COLS if c not in ["cust_id","date"]]

def _agg_safe(s, fn):
    s = pd.to_numeric(s, errors="coerce"); s = s[np.isfinite(s)]
    if s.size == 0: return 0.0
    if fn=="sum":  return float(s.sum())
    if fn=="mean": return float(s.mean())
    if fn=="std":  return float(s.std(ddof=0))
    if fn=="max":  return float(s.max())
    return 0.0

def _months_active(frame: pd.DataFrame) -> int:
    if frame.empty: return 0
    cond = (pd.to_numeric(frame.get("cc_transaction_all_cnt", 0), errors="coerce").fillna(0) > 0) | \
           (pd.to_numeric(frame.get("mobile_eft_all_cnt", 0), errors="coerce").fillna(0) > 0)
    return int(cond.sum())

def _ewm_last(series: pd.Series, span: int):
    series = pd.to_numeric(series, errors="coerce").fillna(0.0)
    if series.empty: return 0.0
    return float(series.ewm(span=span, adjust=False).mean().iloc[-1])

def _slope_last(x: pd.Series):
    """
    Basit OLS eğimi: t=0..n-1, slope = cov(t,x)/var(t)
    """
    x = pd.to_numeric(x, errors="coerce").fillna(0.0)
    n = len(x)
    if n < 2: return 0.0
    t = np.arange(n, dtype=float)
    t_m = t.mean(); x_m = x.mean()
    denom = np.sum((t - t_m)**2)
    if denom <= 0: return 0.0
    num = np.sum((t - t_m) * (x - x_m))
    return float(num / denom)

def aggregate_for_one(h_cust: pd.DataFrame, ref_date: pd.Timestamp):
    feat = {"ref_date": ref_date}
    h = h_cust[h_cust["date"] < ref_date]

    if h.empty:
        for col in BASE_NUM_COLS:
            for w in WINDOWS:
                feat[f"{col}_L{w}M_sum"]=feat[f"{col}_L{w}M_mean"]=feat[f"{col}_L{w}M_std"]=feat[f"{col}_L{w}M_max"]=0.0
        feat["recency_days"]=9999
        for w in WINDOWS:
            feat[f"months_active_L{w}M"]=0.0
            feat[f"activity_rate_L{w}M"]=0.0
            feat[f"inactivity_ratio_L{w}M"]=1.0
            feat[f"cc_share_amt_L{w}M"]=0.0
            feat[f"eft_share_amt_L{w}M"]=0.0
        # EWM spans
        for sp in [3,6,12]:
            feat[f"cc_amt_ewm_span{sp}"]=0.0; feat[f"eft_amt_ewm_span{sp}"]=0.0
            feat[f"cc_cnt_ewm_span{sp}"]=0.0; feat[f"eft_cnt_ewm_span{sp}"]=0.0
        # last & zero months
        feat["last_active_days"]=feat["last_cc_days"]=feat["last_eft_days"]=9999
        for zz in [3,6,12]:
            feat[f"zero_cc_months_L{zz}"]=float(zz); feat[f"zero_eft_months_L{zz}"]=float(zz)
        # slopes
        for w in [6,12]:
            for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt"]:
                feat[f"{base}_slope_L{w}M"]=0.0
    else:
        feat["recency_days"] = int((ref_date - h["date"].max()).days)

        for w in WINDOWS:
            start = ref_date - pd.DateOffset(months=w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            for col in BASE_NUM_COLS:
                s = pd.to_numeric(hw.get(col, 0), errors="coerce").fillna(0.0)
                feat[f"{col}_L{w}M_sum"]  = _agg_safe(s, "sum")
                feat[f"{col}_L{w}M_mean"] = _agg_safe(s, "mean")
                feat[f"{col}_L{w}M_std"]  = _agg_safe(s, "std")
                feat[f"{col}_L{w}M_max"]  = _agg_safe(s, "max")
            ma = float(_months_active(hw))
            feat[f"months_active_L{w}M"]=ma
            feat[f"activity_rate_L{w}M"]=ma/float(w)
            feat[f"inactivity_ratio_L{w}M"]=float(max(w-ma,0))/float(w)
            eps=1e-6
            cc_sum = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0)
            eft_sum= feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0)
            tot = cc_sum + eft_sum
            feat[f"cc_share_amt_L{w}M"]=cc_sum/(tot+eps)
            feat[f"eft_share_amt_L{w}M"]=eft_sum/(tot+eps)

            # slopes (6,12)
            if w in (6,12):
                for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt"]:
                    feat[f"{base}_slope_L{w}M"] = _slope_last(pd.to_numeric(hw.get(base,0), errors="coerce").fillna(0.0))

        # EWM spans
        h_sorted = h.sort_values("date")
        for sp in [3,6,12]:
            feat[f"cc_amt_ewm_span{sp}"] = _ewm_last(h_sorted["cc_transaction_all_amt"], sp)
            feat[f"eft_amt_ewm_span{sp}"] = _ewm_last(h_sorted["mobile_eft_all_amt"], sp)
            feat[f"cc_cnt_ewm_span{sp}"]  = _ewm_last(h_sorted["cc_transaction_all_cnt"], sp)
            feat[f"eft_cnt_ewm_span{sp}"] = _ewm_last(h_sorted["mobile_eft_all_cnt"], sp)

        # last active
        def _last_days(mask):
            if mask.any():
                last_dt = h.loc[mask, "date"].max()
                return int((ref_date - last_dt).days)
            return 9999
        cc_cnt = pd.to_numeric(h.get("cc_transaction_all_cnt", 0), errors="coerce").fillna(0)
        eft_cnt= pd.to_numeric(h.get("mobile_eft_all_cnt", 0), errors="coerce").fillna(0)
        any_act = (cc_cnt>0) | (eft_cnt>0)
        feat["last_active_days"]=_last_days(any_act)
        feat["last_cc_days"]=_last_days(cc_cnt>0)
        feat["last_eft_days"]=_last_days(eft_cnt>0)

        # zero months (3/6/12)
        def _zero_months(last_w):
            start = ref_date - pd.DateOffset(months=last_w)
            hw = h[(h["date"] >= start) & (h["date"] < ref_date)]
            ccz = (pd.to_numeric(hw.get("cc_transaction_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            eftz= (pd.to_numeric(hw.get("mobile_eft_all_cnt",0), errors="coerce").fillna(0)==0).sum()
            return float(ccz), float(eftz)
        for zz in [3,6,12]:
            zc, ze = _zero_months(zz)
            feat[f"zero_cc_months_L{zz}"]=zc; feat[f"zero_eft_months_L{zz}"]=ze

    # per-txn
    eps=1e-6
    for w in WINDOWS:
        feat[f"cc_amt_per_txn_L{w}M"] = feat.get(f"cc_transaction_all_amt_L{w}M_sum",0.0)/(feat.get(f"cc_transaction_all_cnt_L{w}M_sum",0.0)+eps)
        feat[f"mobile_eft_amt_per_txn_L{w}M"] = feat.get(f"mobile_eft_all_amt_L{w}M_sum",0.0)/(feat.get(f"mobile_eft_all_cnt_L{w}M_sum",0.0)+eps)

    # trend diff/ratio
    for short,long in [(1,3),(3,6),(6,12)]:
        for base in ["cc_transaction_all_amt","cc_transaction_all_cnt","mobile_eft_all_amt","mobile_eft_all_cnt","active_product_category_nbr"]:
            ms = feat.get(f"{base}_L{short}M_mean",0.0)
            ml = feat.get(f"{base}_L{long}M_mean",0.0)
            feat[f"{base}_trend_mean_L{short}vL{long}_diff"]=ms-ml
            feat[f"{base}_trend_mean_L{short}vL{long}_ratio"]=ms/(ml+1e-6)

    # HHI & CV
    for w in WINDOWS:
        cc_share  = feat.get(f"cc_share_amt_L{w}M",0.0)
        eft_share = feat.get(f"eft_share_amt_L{w}M",0.0)
        feat[f"channel_hhi_amt_L{w}M"]=cc_share**2 + eft_share**2

    for base in ["cc_transaction_all_amt","mobile_eft_all_amt","cc_transaction_all_cnt","mobile_eft_all_cnt"]:
        for w in [3,6,12]:
            stdv = feat.get(f"{base}_L{w}M_std",0.0)
            mean = feat.get(f"{base}_L{w}M_mean",0.0)
            feat[f"{base}_cv_L{w}M"]= stdv/(mean+eps)

    # recency transform
    feat["recency_log"]   = float(np.log1p(feat.get("recency_days",9999)))
    feat["recency_score"] = float(np.exp(-feat.get("recency_days",9999)/30.0))

    return feat

def build_features_for_keys(hist_df, keys_df, progress_every=4000):
    rows, cache = [], {}
    for i, r in enumerate(keys_df.itertuples(index=False), 1):
        cid, rd = r.cust_id, r.ref_date
        if pd.isna(cid) or pd.isna(rd): continue
        h_cust = cache.get(cid)
        if h_cust is None:
            h_cust = hist_df[hist_df["cust_id"] == cid]
            cache[cid] = h_cust
        f = aggregate_for_one(h_cust, rd); f["cust_id"]=cid; rows.append(f)
        if progress_every and (i % progress_every == 0):
            print(f"[features] {i}/{len(keys_df)}")
    return pd.DataFrame(rows)

# --------------------
# 3) Metrics
# --------------------
def gini_from_auc(y_true, y_score):
    return 2 * roc_auc_score(y_true, y_score) - 1

def recall_at_k(y_true, y_score, k=0.10):
    y_true = np.array(y_true)
    topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]
    return float(y_true[idx].sum()) / float(y_true.sum() + 1e-9)

def lift_at_k(y_true, y_score, k=0.10):
    y_true = np.array(y_true)
    topk = int(np.ceil(len(y_true)*k))
    idx = np.argsort(-y_score)[:topk]
    prec=float(y_true[idx].mean()); prev=float(y_true.mean())
    return prec/max(prev,1e-9)

def competition_score(y_true, y_score):
    g=gini_from_auc(y_true,y_score); r=recall_at_k(y_true,y_score); l=lift_at_k(y_true,y_score)
    return {"Gini":round(g,4),"Recall@10%":round(r,4),"Lift@10%":round(l,4),"CompetitionScore":round(0.4*g+0.3*r+0.3*l,4)}

# --------------------
# 4) Prep for model (native categorical)
# --------------------
def add_calendar_and_micro(df):
    df["ref_month"]   = df["ref_date"].dt.month.astype(int)
    df["ref_quarter"] = df["ref_date"].dt.quarter.astype(int)
    df["ref_month_sin"] = np.sin(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["ref_month_cos"] = np.cos(2*np.pi*df["ref_month"]/12.0).astype(float)
    df["is_year_start_end"] = ((df["ref_month"]==12) | (df["ref_month"]==1)).astype("Int8")
    # mikro etkileşim:
    if "age" in df.columns and "tenure" in df.columns:
        a = pd.to_numeric(df["age"], errors="coerce").fillna(0).astype(float)
        t = pd.to_numeric(df["tenure"], errors="coerce").fillna(0).astype(float)
        df["age_tenure"] = a * t
        df["tenure_log"] = np.log1p(t)
    else:
        df["age_tenure"] = 0.0
        df["tenure_log"] = 0.0
    return df

def prepare_frame_for_model(X):
    X = X.copy()
    num_cols = X.select_dtypes(include="number").columns.tolist()
    cat_cols = [c for c in X.columns if c not in num_cols and c != "ref_date"]

    X[num_cols] = X[num_cols].fillna(0)
    for c in cat_cols:
        X[c] = X[c].astype("category")
        if "Unknown" not in list(X[c].cat.categories):
            X[c] = X[c].cat.add_categories("Unknown")
        X[c] = X[c].fillna("Unknown")

    X_enc = X.drop(columns=["ref_date"])
    feat_cols = [c for c in X_enc.columns if c not in ["cust_id","churn"]]
    cat_feats = [c for c in feat_cols if str(X_enc[c].dtype)=="category"]
    return X_enc, feat_cols, cat_feats

def pick_keys_recent(ref_df, sample_n, recent_frac):
    keys = ref_df[["cust_id","ref_date","churn"]].copy().sort_values("ref_date").reset_index(drop=True)
    q = keys["ref_date"].quantile(1.0 - recent_frac)
    keys = keys[keys["ref_date"] >= q].reset_index(drop=True)
    if len(keys) > sample_n:
        idx = np.linspace(0, len(keys)-1, sample_n).round().astype(int)
        keys = keys.iloc[idx].reset_index(drop=True)
    return keys

# --------------------
# 5) Time-aware CV folds
# --------------------
def make_time_folds(df_dates: pd.Series, qs=VALID_QS):
    """
    Rolling origin: for each q in qs, train on < q, validate on [q, q+Δ)
    Δ = next quantile - q (son q için üst aralık sonuna kadar).
    """
    cuts = [df_dates.quantile(q) for q in qs]
    cuts_sorted = sorted(list(set(cuts)))
    bounds = cuts_sorted + [df_dates.max() + pd.Timedelta(days=1)]
    folds = []
    for i, qcut in enumerate(cuts_sorted):
        train_mask = df_dates < qcut
        val_mask   = (df_dates >= qcut) & (df_dates < bounds[i+1])
        if val_mask.sum() > 0 and train_mask.sum() > 0:
            folds.append((train_mask, val_mask))
    return folds

# --------------------
# 6) Tuning (Optuna veya mikro grid)
# --------------------
def tune_lgbm_timecv(X, y, feat_cols, cat_feats, ref_dates):
    try:
        import optuna
        def objective(trial):
            params = dict(
                objective="binary",
                n_estimators=BASE_N_EST,
                learning_rate=trial.suggest_float("learning_rate", 0.02, 0.05),
                max_depth=trial.suggest_int("max_depth", 6, 12),
                num_leaves=trial.suggest_int("num_leaves", 96, 256),
                min_child_samples=trial.suggest_int("min_child_samples", 15, 60),
                subsample=trial.suggest_float("subsample", 0.7, 1.0),
                subsample_freq=1,
                colsample_bytree=trial.suggest_float("colsample_bytree", 0.6, 1.0),
                reg_alpha=trial.suggest_float("reg_alpha", 0.0, 0.3),
                reg_lambda=trial.suggest_float("reg_lambda", 0.0, 2.0),
                min_split_gain=trial.suggest_float("min_split_gain", 0.0, 0.05),
                max_bin=trial.suggest_int("max_bin", 255, 511),
                random_state=SEED
            )

            # weight modu seçimi
            mode = trial.suggest_categorical("weight_mode", ["balanced","spw"])
            scores = []
            folds = make_time_folds(ref_dates, qs=VALID_QS)
            for (tr, va) in folds:
                X_tr, y_tr = X.loc[tr, feat_cols], y.loc[tr].astype(int)
                X_va, y_va = X.loc[va, feat_cols], y.loc[va].astype(int)

                if mode=="spw":
                    pos = int(y_tr.sum()); neg = int(len(y_tr) - pos)
                    spw = float(neg / max(pos, 1))
                    params_fold = {**params, "scale_pos_weight": spw}
                    params_fold.pop("class_weight", None)
                else:
                    params_fold = {**params, "class_weight": "balanced"}
                    params_fold.pop("scale_pos_weight", None)

                clf = LGBMClassifier(**params_fold)
                clf.fit(
                    X_tr, y_tr,
                    eval_set=[(X_va, y_va)],
                    eval_metric="auc",
                    callbacks=[early_stopping(stopping_rounds=EARLY_STOP_ROUNDS), log_evaluation(period=100)],
                    categorical_feature=cat_feats
                )
                proba = clf.predict_proba(X_va)[:,1]
                met = competition_score(y_va, proba)
                scores.append(met["CompetitionScore"])

            return float(np.mean(scores))

        study = optuna.create_study(direction="maximize")
        study.optimize(objective, n_trials=25, show_progress_bar=False)
        best_params = study.best_params
        # params rebuild:
        final = dict(
            objective="binary",
            n_estimators=BASE_N_EST,
            learning_rate=best_params["learning_rate"],
            max_depth=best_params["max_depth"],
            num_leaves=best_params["num_leaves"],
            min_child_samples=best_params["min_child_samples"],
            subsample=best_params["subsample"],
            subsample_freq=1,
            colsample_bytree=best_params["colsample_bytree"],
            reg_alpha=best_params["reg_alpha"],
            reg_lambda=best_params["reg_lambda"],
            min_split_gain=best_params["min_split_gain"],
            max_bin=best_params["max_bin"],
            random_state=SEED
        )
        if best_params["weight_mode"]=="spw":
            final["scale_pos_weight"] = float((y.size - y.sum()) / max(y.sum(),1))
        else:
            final["class_weight"] = "balanced"
        print("✅ Optuna BEST:", final)
        return final

    except Exception as e:
        print("⚠️ Optuna yok/başarısız:", e, "→ mikro grid'e geçiliyor.")
        # Mikro grid (dar)
        grid = [
            dict(learning_rate=0.03, max_depth=8,  num_leaves=128, min_child_samples=25, subsample=0.9, colsample_bytree=0.85, reg_alpha=0.10, reg_lambda=1.0, max_bin=511, mode="balanced"),
            dict(learning_rate=0.03, max_depth=9,  num_leaves=160, min_child_samples=22, subsample=0.9, colsample_bytree=0.85, reg_alpha=0.05, reg_lambda=1.2, max_bin=511, mode="balanced"),
            dict(learning_rate=0.028,max_depth=9,  num_leaves=160, min_child_samples=20, subsample=0.95,colsample_bytree=0.90, reg_alpha=0.0,  reg_lambda=1.5, max_bin=511, mode="spw"),
            dict(learning_rate=0.032,max_depth=8,  num_leaves=144, min_child_samples=30, subsample=0.85,colsample_bytree=0.80, reg_alpha=0.10, reg_lambda=0.8, max_bin=511, mode="balanced"),
            dict(learning_rate=0.026,max_depth=10, num_leaves=192, min_child_samples=25, subsample=0.90,colsample_bytree=0.85, reg_alpha=0.05, reg_lambda=1.2, max_bin=511, mode="spw"),
        ]
        best, bestscore = None, -1e9
        folds = make_time_folds(ref_dates, qs=VALID_QS)
        for g in grid:
            params = dict(
                objective="binary", n_estimators=BASE_N_EST, subsample_freq=1, random_state=SEED,
                learning_rate=g["learning_rate"], max_depth=g["max_depth"], num_leaves=g["num_leaves"],
                min_child_samples=g["min_child_samples"], subsample=g["subsample"],
                colsample_bytree=g["colsample_bytree"], reg_alpha=g["reg_alpha"], reg_lambda=g["reg_lambda"],
                max_bin=g["max_bin"]
            )
            if g["mode"]=="spw":
                spw = float((y.size - y.sum()) / max(y.sum(), 1))
                params["scale_pos_weight"]=spw
            else:
                params["class_weight"]="balanced"

            scores=[]
            for (tr,va) in folds:
                X_tr, y_tr = X.loc[tr, feat_cols], y.loc[tr].astype(int)
                X_va, y_va = X.loc[va, feat_cols], y.loc[va].astype(int)
                clf = LGBMClassifier(**params)
                clf.fit(
                    X_tr, y_tr,
                    eval_set=[(X_va, y_va)],
                    eval_metric="auc",
                    callbacks=[early_stopping(stopping_rounds=EARLY_STOP_ROUNDS), log_evaluation(period=100)],
                    categorical_feature=cat_feats
                )
                proba = clf.predict_proba(X_va)[:,1]
                scores.append(competition_score(y_va, proba)["CompetitionScore"])
            sc = float(np.mean(scores))
            if sc > bestscore:
                bestscore, best = sc, params
        print("✅ Mikro-grid BEST:", best, "score≈", round(bestscore,4))
        return best

# --------------------
# 7) MAIN
# --------------------
def main():
    # --- TUNING SET (opsiyonel) ---
    if DO_TUNE:
        keys_tune = pick_keys_recent(ref, TUNE_SAMPLE_N, TUNE_RECENT_FRAC)
        print(f"🎯 TUNE keys: {len(keys_tune)}")
        feats_tune = build_features_for_keys(hist, keys_tune[["cust_id","ref_date"]], progress_every=4000)
        X_tune = keys_tune.merge(feats_tune, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
        X_tune = add_calendar_and_micro(X_tune)
        X_enc, feat_cols, cat_feats = prepare_frame_for_model(X_tune)

        # low-variance (numeric)
        nunique = X_enc[feat_cols].nunique(dropna=False)
        lowvar = nunique[nunique<=1].index.tolist()
        if len(lowvar):
            print(f"🧹 (TUNE) Low-var drop: {len(lowvar)}")
            feat_cols = [c for c in feat_cols if c not in lowvar]

        # Tuning
        best_params = tune_lgbm_timecv(X_enc, X_tune["churn"], feat_cols, cat_feats, X_tune["ref_date"])
    else:
        # Güçlü bir baz
        best_params = dict(
            objective="binary", n_estimators=BASE_N_EST, learning_rate=0.03,
            max_depth=8, num_leaves=128, min_child_samples=25,
            subsample=0.90, subsample_freq=1, colsample_bytree=0.85,
            reg_alpha=0.10, reg_lambda=1.0, max_bin=511,
            class_weight="balanced", random_state=SEED
        )

    # --- FULL TRAIN ---
    keys_full = ref[["cust_id","ref_date","churn"]].copy().sort_values("ref_date").reset_index(drop=True)
    if TRAIN_RECENT_FRAC_FULL < 1.0:
        q = keys_full["ref_date"].quantile(1.0 - TRAIN_RECENT_FRAC_FULL)
        keys_full = keys_full[keys_full["ref_date"] >= q].reset_index(drop=True)
    print(f"🧩 FULL TRAIN keys: {len(keys_full)}")

    feats_full = build_features_for_keys(hist, keys_full[["cust_id","ref_date"]], progress_every=6000)
    X_full = keys_full.merge(feats_full, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
    X_full = add_calendar_and_micro(X_full)
    X_enc, feat_cols, cat_feats = prepare_frame_for_model(X_full)

    # low-var
    nunique = X_enc[feat_cols].nunique(dropna=False)
    lowvar = nunique[nunique<=1].index.tolist()
    if len(lowvar):
        print(f"🧹 (FULL) Low-var drop: {len(lowvar)}")
        feat_cols = [c for c in feat_cols if c not in lowvar]

    TARGET="churn"
    clf = LGBMClassifier(**best_params)
    # full train — early stopping yok (valid yok); yine de overfit’i sınırlayan paramlar var
    clf.fit(X_enc[feat_cols], X_full[TARGET].astype(int), categorical_feature=cat_feats)
    print("✅ FULL TRAIN tamam.")

    # --- SUBMISSION SET ---
    if SUB_KEYS is not None:
        sub_keys = pd.read_csv(SUB_KEYS, parse_dates=["ref_date"], low_memory=False)
        if "cust_id" in sub_keys.columns:
            sub_keys["cust_id"] = pd.to_numeric(sub_keys["cust_id"], errors="coerce").astype("Int64")
        sub_keys = sub_keys[["cust_id","ref_date"]]
        print("📦 Submission keys:", sub_keys.shape)
    else:
        print("⚠️ Ayrı submission keys bulunamadı — ref üzerinde skor üreteceğim.")
        sub_keys = ref[["cust_id","ref_date"]].copy()

    feats_sub = build_features_for_keys(hist, sub_keys[["cust_id","ref_date"]], progress_every=6000)
    X_sub = sub_keys.merge(feats_sub, on=["cust_id","ref_date"], how="left").merge(cust, on="cust_id", how="left")
    X_sub = add_calendar_and_micro(X_sub)

    # train kategorileri ile hizala
    X_sub_enc = X_sub.copy()
    num_cols = X_sub_enc.select_dtypes(include="number").columns.tolist()
    cat_cols = [c for c in X_sub_enc.columns if c not in num_cols and c != "ref_date"]
    X_sub_enc[num_cols] = X_sub_enc[num_cols].fillna(0)
    for c in cat_cols:
        X_sub_enc[c] = X_sub_enc[c].astype("category")
        if "Unknown" not in list(X_sub_enc[c].cat.categories):
            X_sub_enc[c] = X_sub_enc[c].cat.add_categories("Unknown")
        X_sub_enc[c] = X_sub_enc[c].fillna("Unknown")
        # train'in kategorilerini kullan
        if c in X_enc.columns and str(X_enc[c].dtype)=="category":
            X_sub_enc[c] = X_sub_enc[c].cat.set_categories(X_enc[c].cat.categories)

    X_sub_mat = X_sub_enc[feat_cols]
    proba = clf.predict_proba(X_sub_mat)[:,1]
    out = X_sub[["cust_id","ref_date"]].copy()
    out["score"] = proba
    out["ref_date"] = pd.to_datetime(out["ref_date"]).dt.strftime("%Y-%m-%d")
    out.to_csv(SUBMISSION_PATH, index=False)
    print(f"💾 submission yazıldı: {SUBMISSION_PATH}")
    print(out.head())

if __name__ == "__main__":
    main()
```
